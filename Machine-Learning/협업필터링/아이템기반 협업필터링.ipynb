```python
import pandas as pd
import numpy as np

movies = pd.read_csv("./data/movies.csv")
ratings = pd.read_csv("./data/ratings.csv")
```

```python
movies
```

```python
ratings
```

```python
# movieID를 기준으로 병합
rating_movie = pd.merge(ratings, movies, on="movieId")

rating_movie
```

```python
# 피벗테이블 생성 (인덱스 : 아이템, 컬럼: 유저)
rating_matrix = rating_movie.pivot_table(
    index = 'title',     # 행 : 영화제목
    columns = 'userId',  # 열 : 사용자 ID
    values = 'rating'    # 값 : 평점
).fillna(0)

rating_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity # cosine 유사도 계산 도구

item_sim = cosine_similarity(
    rating_matrix,
    rating_matrix
) # 영화간에 유사도 계산 (아이템 - 아이템끼리)

item_sim
```

```python
item_sim_df = pd.DataFrame(
    item_sim,                        # 영화끼리의 유사도 계산 배열 데이터
    index = rating_matrix.index,     # 행 : 영화제목
    columns = rating_matrix.index    # 열 : 영화제목
)

item_sim_df  # 영화끼리의 유사도 데이터프레임
```

```python
item_sim_df['Toy Story (1995)'].sort_values(ascending=False)[1:11] # toysstory와 가장 비슷한 영화 10개 (본인을 제외해야해서 1부터 11까지)
```

```python
item_sim_df['Toy Story 2 (1999)'].sort_values(ascending=False)[1:11] # toysstory2와 가장 비슷한 영화 10개 (본인을 제외해야해서 1부터 11까지)
```

```python
# 영화 추천해주는 함수
def movie_recommendations(movie_name, n):
    # 유사도 높은 순으로 정렬 후, 자기 자신을 제외한 상위 n개 출력
    movie_sim = item_sim_df[movie_name].sort_values(ascending=False)[1:n+1]

    return movie_sim.index # 영화 제목만 반환
```
