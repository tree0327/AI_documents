## 1. ÌåÅ(Tips) Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù
Î®ºÏ†Ä `Seaborn` ÎùºÏù¥Î∏åÎü¨Î¶¨Ïóê ÎÇ¥Ïû•Îêú `tips` Îç∞Ïù¥ÌÑ∞Î•º Î∂àÎü¨ÏòµÎãàÎã§.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ Î°úÎìú
df = sns.load_dataset('tips')
df
```

ÏÇ¨ÎûåÎì§Ïù¥ ÌåÅÏùÑ ÏñºÎßàÎÇò ÎÇ¥ÎäîÏßÄ **ÌûàÏä§ÌÜ†Í∑∏Îû®**ÏúºÎ°ú Í∑∏Î†§ÏÑú ÌôïÏù∏Ìï¥Î¥ÖÎãàÎã§.

## 2. Î∂ìÍΩÉ(Iris) Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù
Ïù¥Î≤àÏóêÎäî ÍΩÉÏûéÍ≥º ÍΩÉÎ∞õÏπ®Ïùò Í∏∏Ïù¥Î•º Ï∏°Ï†ïÌïú `iris` Îç∞Ïù¥ÌÑ∞Î•º ÏÇ¥Ìé¥Î≥ºÍπåÏöî?


```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Iris Îç∞Ïù¥ÌÑ∞ÏÖã Î°úÎìú Î∞è DataFrame Î≥ÄÌôò
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df
```

ÍΩÉÏûéÏùò Í∏∏Ïù¥(`petal length`)Í∞Ä Ïñ¥ÎñªÍ≤å ÌçºÏ†∏ÏûàÎäîÏßÄ ÌûàÏä§ÌÜ†Í∑∏Îû®ÏúºÎ°ú ÌôïÏù∏Ìï©ÎãàÎã§.


```python
# ÍΩÉÏûé Í∏∏Ïù¥Ïóê ÎåÄÌïú ÌûàÏä§ÌÜ†Í∑∏Îû®
plt.figure(figsize=(8, 5))

sns.histplot(df['petal length (cm)'], kde=True)
plt.title('Petal Length Distribution')
plt.show()
```

### üî• ÌäπÏÑ± Í∞ÑÏùò Í¥ÄÍ≥Ñ (ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ)
ÍΩÉÎ∞õÏπ®Í≥º ÍΩÉÏûéÏùò Í∏∏Ïù¥, ÎÑàÎπÑÍ∞Ä ÏÑúÎ°ú Ïñ¥Îñ§ Í¥ÄÍ≥ÑÍ∞Ä ÏûàÎäîÏßÄ **ÌûàÌä∏Îßµ**ÏúºÎ°ú Í∑∏Î†§Î¥ÖÎãàÎã§.
ÏÉâÏù¥ Î∞ùÏùÑÏàòÎ°ù(ÎòêÎäî ÏÑ§Ï†ïÏóê Îî∞Îùº ÏßÑÌï†ÏàòÎ°ù) Í¥ÄÍ≥ÑÍ∞Ä ÍπäÎã§Îäî ÎúªÏù¥ÏóêÏöî.


```python
# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌûàÌä∏Îßµ
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## 3. ÏôÄÏù∏(Wine) Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù
`Scikit-learn`ÏóêÏÑú Ï†úÍ≥µÌïòÎäî ÏôÄÏù∏ Îç∞Ïù¥ÌÑ∞Î•º Î∂àÎü¨ÏôÄÏÑú Ï¢Ä Îçî Î≥µÏû°Ìïú Í¥ÄÍ≥ÑÎ•º ÏÇ¥Ìé¥Î¥ÖÏãúÎã§.


```python
# Scikit-learnÏùò Wine Îç∞Ïù¥ÌÑ∞ÏÖã Î°úÎìú
from sklearn.datasets import load_wine
wine = load_wine()
wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_df
```

### üß© Î™®Îì† ÌäπÏÑ± ÌïúÎààÏóê Î≥¥Í∏∞ (Pairplot)
Ïó¨Îü¨ Îç∞Ïù¥ÌÑ∞ÎÅºÎ¶¨Ïùò Í¥ÄÍ≥ÑÎ•º Ìïú Î≤àÏóê Í∑∏ÎûòÌîÑÎ°ú Í∑∏Î†§Ï£ºÎäî `pairplot`ÏûÖÎãàÎã§.
Îç∞Ïù¥ÌÑ∞Í∞Ä Ïñ¥ÎñªÍ≤å ÎÇòÎâòÏñ¥ ÏûàÎäîÏßÄ Ï†ÑÏ≤¥Ï†ÅÏù∏ Í∑∏Î¶ºÏùÑ Î≥º Îïå ÏïÑÏ£º Ïú†Ïö©Ìï¥Ïöî.


```python
# ÏôÄÏù∏ Îç∞Ïù¥ÌÑ∞ÏÖãÏùò Î∂ÑÌè¨ ÏãúÍ∞ÅÌôî
sns.pairplot(wine_df)
plt.show()
```

## 4. Ïù∏ÌÑ∞ÎÑ∑ÏóêÏÑú Îç∞Ïù¥ÌÑ∞ Í∞ÄÏ†∏Ïò§Í∏∞
ÏõπÏÇ¨Ïù¥Ìä∏ Ï£ºÏÜå(URL)Î•º Ïù¥Ïö©Ìï¥ÏÑú Îç∞Ïù¥ÌÑ∞Î•º ÏßÅÏ†ë Î∂àÎü¨Ïò§Îäî Î∞©Î≤ïÏûÖÎãàÎã§.


```python
import numpy as np
import pandas as pd

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(
        data_url,
        sep="\s+",
        skiprows=22,
        header=None
    )
data = np.hstack(
        [raw_df.values[::2, :], raw_df.values[1::2, :3]]
    )

column_names = [
        'CRIM',
        'ZN',
        'INDUS',
        'CHAS',
        'NOX',
        'RM',
        'AGE',
        'DIS',
        'RAD',
        'TAX',
        'PTRATIO',
        'B',
        'LSTAT',
        'MEDV'
    ]
boston_df = pd.DataFrame(data, columns=column_names)

# Í≤∞Ï∏°Í∞í Ï∂îÍ∞Ä (ÏòàÏãúÏö©)
boston_df.loc[5:10, 'CRIM'] = np.nan

# Í≤∞Ï∏°Í∞í Ï≤òÎ¶¨: ÌèâÍ∑†Í∞íÏúºÎ°ú ÎåÄÏ≤¥
boston_df['CRIM'].fillna(boston_df['CRIM'].mean(), inplace=True)

# Ïù¥ÏÉÅÏπò ÌÉêÏßÄ: Î∞ïÏä§ÌîåÎ°ØÏúºÎ°ú ÏãúÍ∞ÅÌôî
plt.figure(figsize=(8, 5))
sns.boxplot(x=boston_df['CRIM'])
plt.title('CRIM Feature Outliers Detection')
plt.show()
```

### üìè Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî (Preprocessing)
Îç∞Ïù¥ÌÑ∞Îì§Ïùò ÌÅ¨Í∏∞(Îã®ÏúÑ)Í∞Ä Îã§Î•¥Î©¥ Î∂ÑÏÑùÏù¥ Ïñ¥Î†§Ïö∏ Ïàò ÏûàÏñ¥Ïöî.
`StandardScaler`Î•º ÏÇ¨Ïö©Ìï¥ÏÑú Îç∞Ïù¥ÌÑ∞Î•º ÎπÑÏä∑Ìïú Î≤îÏúÑÎ°ú ÎßûÏ∂∞Ï£ºÎäî(ÌëúÏ§ÄÌôî) ÏûëÏóÖÏûÖÎãàÎã§.


```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Ïä§ÏºÄÏùºÎßÅ: ÌëúÏ§ÄÌôî
scaler = StandardScaler()
boston_df[['CRIM', 'ZN']] = scaler.fit_transform(boston_df[['CRIM', 'ZN']])

# Ï†ïÍ∑úÌôî
scaler = MinMaxScaler()
boston_df[['AGE', 'TAX']] = scaler.fit_transform(boston_df[['AGE', 'TAX']])
```
