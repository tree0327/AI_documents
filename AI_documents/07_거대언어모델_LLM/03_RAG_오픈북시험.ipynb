# 03. AI에게 컨닝 페이퍼 주기: RAG (검색 증강 생성) 📚

> **목표**: AI가 모르는 정보(최신 뉴스, 사내 문서)를 물어보면 어떻게 대답하게 할까요? AI에게 **참고서**를 쥐어주는 기술인 **RAG(Retrieval-Augmented Generation)**를 알아봅시다.

## 1. LLM의 한계: 닫힌 책 시험 (Closed Book Exam)
ChatGPT 같은 모델은 학습이 끝난 시점(예: 2023년) 이후의 일은 모릅니다. 
마치 수능날, **머릿속에 외운 지식만 가지고** 시험을 보는 것과 같습니다.

- **문제점**:
    1. **최신 정보 부재**: "어제 삼성전자 주가 얼마야?" -> 모름.
    2. **할루시네이션 (거짓말)**: 모르는 것도 아는 척 지어냄.

## 2. RAG: 오픈 북 시험 (Open Book Exam)
그래서 우리는 AI에게 **검색(Retrieval)** 능력을 달아줍니다. 질문이 들어오면, 관련 문서를 찾아서 보여주고 답하게 하는 것입니다.

**RAG의 3단계**:
1.  **검색 (Retrieval)**: 사용자 질문과 관련된 문서를 데이터베이스(벡터 DB)에서 찾아옵니다. (도서관에서 책 찾기)
2.  **증강 (Augmented)**: 찾은 정보를 프롬프트에 끼워 넣습니다. (컨닝 페이퍼 만들기)
    > "사용자가 '사내 복지'를 물어봤어. 여기 '사내 복지 규정.pdf' 내용이 있으니까 이거 읽고 대답해."
3.  **생성 (Generation)**: AI가 그 정보를 바탕으로 답변을 작성합니다.

## 3. RAG가 중요한 이유
기업에서 AI를 도입할 때 **가장 필수적인 기술**입니다.
- 우리 회사의 비밀 문서나 매뉴얼을 AI에게 학습(Fine-tuning)시키는 건 너무 비싸고 오래 걸립니다.
- 대신 RAG를 쓰면, 문서를 폴더에 넣기만 해도 AI가 실시간으로 읽고 대답해줍니다.
