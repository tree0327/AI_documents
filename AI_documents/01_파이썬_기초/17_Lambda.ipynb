```python
# Lambda: 호출될 때 값이 구해지는 하나의 표현식
# 익명 함수(anonymous function)
lambda param: param**2
# :의 왼편에 오는 부분이 파라미터를 정의하는 부분
# :의 오른편에 오는 부분이 expresion이며 반환되는 값. 
# 값이 하나여야 함. 
# inline으로 표현 가능해야함
```

```python
some_func = lambda param: param**2
```

```python
some_func(5)
```

```python
# map(func, *iterables) 내장 함수
help(map)
```

```python
a = list(range(6))
print(a)
```

```python
b = [_var * 2 for _var in a]
print(b)
```

```python
map_result = map(lambda x, y: x + y, a, b)
```

```python
print(map_result)
```

```python
list(map_result)
```
