# 01. Python이란 무엇인가? (feat. 통역사)

> **목표**: 프로그래밍 언어의 개념을 이해하고, 파이썬이 왜 AI 배우기에 가장 좋은 언어인지 알아봅시다.

## 1. 프로그래밍 언어는 '통역사'다
우리가 외국인 친구와 대화하려면 '영어'나 '한국어' 같은 언어가 필요하죠? 
컴퓨터도 마찬가지입니다. 컴퓨터는 0과 1(이진수)밖에 모르는 바보 기계라서, 우리가 하는 말을 알아듣지 못해요.

그래서 우리(인간)의 말을 컴퓨터가 알아들을 수 있는 0과 1로 바꿔주는 **통역사**가 필요한데, 이게 바로 **프로그래밍 언어(Python, C, Java 등)**입니다.

- **기계어 (Machine Language)**: 컴퓨터가 쓰는 0과 1 (`01010101`) -> 인간이 배우기엔 너무 어려움
- **고급 언어 (High-Level Language)**: 인간이 쓰는 영어와 비슷함 (`print("Hello")`) -> **파이썬(Python)**이 여기 속함!

> [!TIP]
> **파이썬이 쉬운 이유?**
> 파이썬은 문법이 아주 직관적입니다. 마치 영어 문장을 쓰는 것처럼 코드를 짤 수 있어서, 초보자가 배우기에 가장 적합해요.

## 2. Python의 특징 (인터프리터 언어)
파이썬은 **인터프리터(Interpreter)** 언어라고 해요. 어려운 용어 같지만, **'동시 통역'**이라고 생각하면 됩니다.

- **컴파일러 언어 (C, Java)**: 책 한 권을 통째로 번역한 뒤에 보여줍니다. (속도는 빠르지만, 번역 시간이 걸림)
- **인터프리터 언어 (Python)**: 한 줄씩 실시간으로 통역해서 실행합니다. (결과를 바로바로 볼 수 있어서 편함!)
