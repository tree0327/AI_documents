**ë¹„ìœ **: 
- ì—¬ëŸ¬ë¶„ì´ ì¹œêµ¬ì—ê²Œ ì¤„ ì„ ë¬¼ì„ ìƒ€ìŠµë‹ˆë‹¤. (**ê¸°ë³¸ í•¨ìˆ˜**)

- ë‚´ìš©ë¬¼ì€ ë³€í•˜ì§€ ì•Šì•˜ì§€ë§Œ, 'í¬ì¥'ì´ë¼ëŠ” ê¸°ëŠ¥ì´ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.

````python
# ë°ì½”ë ˆì´í„° ì •ì˜í•˜ê¸° (í¬ì¥ì§€ ë§Œë“¤ê¸°)
def gift_wrap(func):
    def wrapper():
        print("ğŸ í¬ì¥ì„ ëœ¯ìŠµë‹ˆë‹¤...")
        func()  # ì›ë˜ ì„ ë¬¼(í•¨ìˆ˜) ì‹¤í–‰
        print("âœ¨ ì§œì”!")
    return wrapper

# ë°ì½”ë ˆì´í„° ì‚¬ìš©í•˜ê¸° (í¬ì¥ì§€ë¡œ ì‹¸ê¸°)
@gift_wrap
def give_iphone():
    print("ğŸ“± ì•„ì´í° 15 Pro")

# ì‹¤í–‰í•´ë³´ê¸°
give_iphone()
````

## 2. ì œë„ˆë ˆì´í„° (Generator) - íšŒì „ ì´ˆë°¥ ğŸ£
ë¦¬ìŠ¤íŠ¸(List)ëŠ” ë°ì´í„°ë¥¼ í•œêº¼ë²ˆì— ë‹¤ ë§Œë“¤ì–´ì„œ ë©”ëª¨ë¦¬ì— ì˜¬ë ¤ë‘ì§€ë§Œ, ì œë„ˆë ˆì´í„°ëŠ” **í•„ìš”í•  ë•Œë§ˆë‹¤ í•˜ë‚˜ì”©** ë§Œë“¤ì–´ëƒ…ë‹ˆë‹¤.

**ë¹„ìœ **:
- **ë¦¬ìŠ¤íŠ¸**: ë·”í˜. ëª¨ë“  ìŒì‹ì´ ì ‘ì‹œì— ë‹´ê²¨ì„œ í…Œì´ë¸” ìœ„ì— ê½‰ ì°¨ ìˆìŠµë‹ˆë‹¤. (ë©”ëª¨ë¦¬ ë§ì´ ì”€)
- **ì œë„ˆë ˆì´í„°**: íšŒì „ ì´ˆë°¥. ë‚´ê°€ ë¨¹ê³  ì‹¶ì„ ë•Œ í•˜ë‚˜ì”© ë ˆì¼ ìœ„ë¡œ ì˜¬ë¼ì˜µë‹ˆë‹¤. (ë©”ëª¨ë¦¬ ì•„ë‚Œ)

í•µì‹¬ í‚¤ì›Œë“œëŠ” `yield` ì…ë‹ˆë‹¤. í•¨ìˆ˜ê°€ `return`ìœ¼ë¡œ ëë‚˜ì§€ ì•Šê³ , `yield`ë¡œ ê°’ì„ í•˜ë‚˜ì”© ì–‘ë³´(ìƒì„±)í•˜ê³  ì ì‹œ ë©ˆì¶¥ë‹ˆë‹¤.

````python
# ì¼ë°˜ì ì¸ ë¦¬ìŠ¤íŠ¸ ë°©ì‹ (ë©”ëª¨ë¦¬ë¥¼ ë§ì´ ì”€)
def make_list_big():
    result = []
    for i in range(3):
        result.append(i * 10)
    return result

print("ë¦¬ìŠ¤íŠ¸:", make_list_big())

# ì œë„ˆë ˆì´í„° ë°©ì‹ (ë©”ëª¨ë¦¬ ì ˆì•½)
def make_generator_big():
    for i in range(3):
        yield i * 10  # í•˜ë‚˜ ë§Œë“¤ê³  ëŒ€ê¸°!

gen = make_generator_big()
print("ì œë„ˆë ˆì´í„°:", gen)
````
