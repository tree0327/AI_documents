```python
# for문
# 반복문
# 코드 블록을 요소의 개수만큼 반복실행

# 만약 Hello world라는 문자열을 10번 출력하고싶다면 어떻게 코드를 작성할것인가?
```

```python
range(10) # 0부터 1씩 증가하는 정수를 10미만일 때까지 만들어준다.
# 즉 range(숫자)는 숫자만큼 요소를 만들어준다.
```

```python
for variable in range(10): # variable은 0부터 9까지 숫자를 차례로 받아온다.
    print("Hello world")
```

```python
# range() 내장함수로 리스트 만들기
# range(시작=0, 끝, 증감=1)
# range(1000) -> 0부터 1씩 증가하는 정수를 만든다. 끝 숫자보다 미만일 경우까지만 동작

# 2부터 100미만에 2씩 증가하는 숫자를 변수에 담기
even_numbers = list(range(2, 100, 2))
```

```python
print(even_numbers) # 2부터 100미만에 2씩 증가하는 숫자를 출력
```

```python
# iterable: 한번에 하나씩 요소를 순차적으로 꺼내올 수 있는 객체
# str, list, tuple, range
for n in [1, 2, 3, 4, 5, 6, 7, 8, 9]:
    print("값은 " + str(n) + "입니다.")
```

```python
for n in reversed("Hello world"):
    print(n)
```

```python
# 시퀀스 자료형의 인덱스를 구해서 그 인덱스로 반복하면서 인덱싱을 하는 예제
exam_list = ["가", "나", "다", "라"]
#             0    1     2    3
# len(exam_list) => 4
# range(4) => 0, 1, 2, 3
for idx in range(len(exam_list)):
    print(exam_list[idx])
```
