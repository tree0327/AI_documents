> **ëª©í‘œ**: ë¨¸ì‹ ëŸ¬ë‹ê³¼ ë”¥ëŸ¬ë‹ í•™ìŠµì˜ í•µì‹¬ ì›ë¦¬ì¸ **ê²½ì‚¬í•˜ê°•ë²•(Gradient Descent)**ì„ ì´í•´í•©ë‹ˆë‹¤. ìˆ˜ì‹ë³´ë‹¤ëŠ” **'ëˆˆì„ ê°€ë¦¬ê³  ì‚° ë‚´ë ¤ì˜¤ê¸°'** ë¹„ìœ ë¡œ ì§ê´€ì ìœ¼ë¡œ ì ‘ê·¼í•´ë´…ì‹œë‹¤.

## 1. ì„ í˜• íšŒê·€ (Linear Regression): ìµœì ì˜ ì„  ê¸‹ê¸°
ìš°ë¦¬ê°€ ì˜ˆì¸¡í•˜ê³  ì‹¶ì€ ë°ì´í„°(ì§‘ê°’, ì ìˆ˜ ë“±)ë¥¼ ê°€ì¥ ì˜ ì„¤ëª…í•˜ëŠ” **ì§ì„  í•˜ë‚˜**ë¥¼ ê¸‹ëŠ” ê²ƒì…ë‹ˆë‹¤.

- **Y = Wx + b**
    - **W (Weight, ê°€ì¤‘ì¹˜)**: ê¸°ìš¸ê¸°. ì…ë ¥ê°’(x)ì´ ì •ë‹µ(Y)ì— ì–¼ë§ˆë‚˜ ì˜í–¥ì„ ë¯¸ì¹˜ëŠ”ì§€.
    - **b (Bias, í¸í–¥)**: ì ˆí¸. ê¸°ë³¸ ë² ì´ìŠ¤ ê°’.

ë”¥ëŸ¬ë‹ì€ ê²°êµ­, ë°ì´í„°ë¥¼ ê°€ì¥ ì˜ ë§íˆëŠ” ìµœì ì˜ **W(ê¸°ìš¸ê¸°)**ì™€ **b(ì ˆí¸)**ë¥¼ ì°¾ëŠ” ê³¼ì •ì…ë‹ˆë‹¤.

## 2. ê²½ì‚¬í•˜ê°•ë²• (Gradient Descent): ì‚°ì—ì„œ ë‚´ë ¤ì˜¤ëŠ” ë°©ë²•
**ë¹„ìœ **:
- ì—¬ëŸ¬ë¶„ì€ ìº„ìº„í•œ ë°¤, ì‚° ê¼­ëŒ€ê¸°ì—ì„œ ëˆˆì´ ê°€ë ¤ì§„ ì±„ë¡œ ì„œ ìˆìŠµë‹ˆë‹¤. (**ì´ˆê¸° ìƒíƒœ**)
- ëª©í‘œëŠ” ì‚° ì•„ë˜ ê°€ì¥ ë‚®ì€ ê³„ê³¡(**ì—ëŸ¬ê°€ ìµœì†Œì¸ ì§€ì **)ìœ¼ë¡œ ë‚´ë ¤ê°€ëŠ” ê²ƒì…ë‹ˆë‹¤.
- ëˆˆì´ ì•ˆ ë³´ì´ë‹ˆê¹Œ, ë°œëì˜ ê°ê°(**ë¯¸ë¶„/ê¸°ìš¸ê¸°**)ë§Œ ë¯¿ê³  ê²½ì‚¬ê°€ ì§„ ìª½ìœ¼ë¡œ í•œ ë°œìêµ­ì”©(**Learning Rate**) ë‚´ë ¤ê°‘ë‹ˆë‹¤.
- **SGD (í™•ë¥ ì  ê²½ì‚¬í•˜ê°•ë²•)**: í•œ ë²ˆì— ë‹¤ ê³„ì‚°í•˜ë©´ ë„ˆë¬´ ì˜¤ë˜ ê±¸ë¦¬ë‹ˆê¹Œ, ë°ì´í„°ë¥¼ ëœë¤ìœ¼ë¡œ ì¡°ê¸ˆì”© ë½‘ì•„ì„œ ë‚´ë ¤ê°€ëŠ” ë°©ì‹ì…ë‹ˆë‹¤. (ë¹ ë¦„!)

## 3. ì‹¤ìŠµ: ìº˜ë¦¬í¬ë‹ˆì•„ ì§‘ê°’ ì˜ˆì¸¡í•˜ê¸°
ì§‘ê°’ ë°ì´í„°ë¥¼ ë³´ê³ , ì§‘ê°’ì„ ì˜ˆì¸¡í•˜ëŠ” ì„ í˜• ëª¨ë¸ì„ ê²½ì‚¬í•˜ê°•ë²•ìœ¼ë¡œ í•™ìŠµì‹œì¼œë´…ì‹œë‹¤.

```python
# í•„ìš”í•œ ë„êµ¬ë“¤ ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import SGDRegressor  # ê²½ì‚¬í•˜ê°•ë²• íšŒê·€ ëª¨ë¸
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler # ë°ì´í„° ìŠ¤ì¼€ì¼ë§ (ì •ê·œí™”)
from sklearn.metrics import mean_squared_error

# 1. ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv("./data/fetch_california_housing.csv")
df = df.rename(columns={"MedHouseVal": "Target"}) # ë³´ê¸° í¸í•˜ê²Œ ì´ë¦„ ë³€ê²½
df.head()
```

## 4. ë°ì´í„° ì‚´í´ë³´ê¸° (ì‹œê°í™”)
ì†Œë“(MedInc)ê³¼ ì§‘ê°’(Target)ì€ ì–´ë–¤ ê´€ê³„ê°€ ìˆì„ê¹Œìš”? ì ì„ ì°ì–´ë³´ë©´ ëŒ€ì¶© ìš°ìƒí–¥í•˜ëŠ” ì§ì„  í˜•íƒœê°€ ë³´ì¼ ê²ƒì…ë‹ˆë‹¤.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="MedInc", y="Target", data=df, alpha=0.5)
plt.title("ì†Œë“ê³¼ ì§‘ê°’ì˜ ê´€ê³„ (Median Income vs Price)")
plt.xlabel("ì†Œë“ (Median Income)")
plt.ylabel("ì§‘ê°’ (House Price)")
plt.show()
```

```python
# 2. ë°ì´í„° ë‚˜ëˆ„ê¸° & ìŠ¤ì¼€ì¼ë§
# ê²½ì‚¬í•˜ê°•ë²•ì€ ìˆ«ìì˜ í¬ê¸°ì— ë¯¼ê°í•´ì„œ, ìŠ¤ì¼€ì¼ë§(ë²”ìœ„ ë§ì¶”ê¸°)ì´ í•„ìˆ˜ì…ë‹ˆë‹¤!
X = df.drop("Target", axis=1)
y = df["Target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("ë°ì´í„° ì¤€ë¹„ ì™„ë£Œ!")
```

```python
# 3. ëª¨ë¸ í•™ìŠµ (ì‚° ë‚´ë ¤ì˜¤ê¸° ì‹œì‘!)
# max_iter=1000 (ìµœëŒ€ 1000ë²ˆ ê±¸ìŒ), eta0=0.01 (ë³´í­)
model = SGDRegressor(max_iter=1000, eta0=0.01, random_state=42)
model.fit(X_train_scaled, y_train)

print("í•™ìŠµ ë! ìµœì ì˜ Wì™€ bë¥¼ ì°¾ì•˜ìŠµë‹ˆë‹¤.")
```

```python
# 4. ê²°ê³¼ í™•ì¸ (ì±„ì )
y_pred = model.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)

print(f"í‰ê·  ì˜¤ì°¨ (MSE): {mse:.4f}")

# ì‹¤ì œ ê°’ vs ì˜ˆì¸¡ ê°’ ì‹œê°í™”
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.3)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2) # ì™„ë²½í•œ ì˜ˆì¸¡ì„ 
plt.xlabel("ì‹¤ì œ ì§‘ê°’")
plt.ylabel("ëª¨ë¸ì´ ì˜ˆì¸¡í•œ ì§‘ê°’")
plt.title("ì •ë‹µ vs ì˜ˆì¸¡ ë¹„êµ")
plt.show()
```
