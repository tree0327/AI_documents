# 01. 텍스트를 숫자로 바꾸기 (Text Vectorization) 🔢

> **목표**: 컴퓨터는 '사과', '바나나' 같은 글자를 이해하지 못합니다. 텍스트를 컴퓨터가 좋아하는 **숫자(벡터)**로 바꾸는 과정인 **'벡터화(Vectorization)'**를 배워봅시다.

## 1. 컴퓨터는 글자를 싫어해
우리가 "나는 밥을 먹었다"라고 말하면 컴퓨터는 이걸 `010101` 같은 숫자로 받아들여야 계산할 수 있습니다.
마치 외국어 사전에서 단어를 찾아 **페이지 번호**로 기억하는 것과 비슷합니다.

- **정수 인코딩 (Integer Encoding)**: 각 단어에 고유한 번호를 붙여주는 것.
    - 사과: 1번, 바나나: 2번, 딸기: 3번...

하지만 이 방식은 문제가 있습니다. 
**1번(사과) + 2번(바나나) = 3번(딸기)?** 
숫자끼리 더하거나 곱했을 때, 엉뚱한 뜻이 되어버릴 수 있죠. 그래서 더 똑똑한 방법이 필요합니다.

## 2. 원-핫 인코딩 (One-Hot Encoding): 메뉴판 고르기

**비유**:
- 식당 메뉴판에 10개의 음식이 있습니다.
- 여러분은 한 번에 **딱 하나**만 주문할 수 있습니다.
- 주문할 음식 칸에만 도장(`1`)을 찍고, 나머지는 모두 비워두는(`0`) 방식입니다.

| 단어 | 사과 | 바나나 | 딸기 |
|:---:|:---:|:---:|:---:|
| **사과** | **1** | 0 | 0 |
| **바나나** | 0 | **1** | 0 |
| **딸기** | 0 | 0 | **1** |

이렇게 하면 단어들이 서로 섞이지 않고 독립적으로 존재하게 됩니다.

## 3. 실습: 원-핫 인코딩 만들어보기

```python
# 사이킷런(Scikit-learn) 라이브러리를 사용해봅시다
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# 1. 데이터 준비 (과일 바구니)
fruits = [["사과"], ["바나나"], ["딸기"], ["사과"]]

# 2. 메뉴판 만들기 (인코더 생성)
encoder = OneHotEncoder()

# 3. 도장 찍기 (변환)
onehot_encoded = encoder.fit_transform(fruits).toarray()

print("원래 데이터:", [f[0] for f in fruits])
print("\n변환된 숫자 (One-Hot Vector):")
print(onehot_encoded)
print("\n메뉴판 순서:", encoder.categories_)
```

> [!TIP]
> **희소 행렬 (Sparse Matrix)**: 원-핫 인코딩은 대부분의 값이 0인 행렬을 만듭니다. 단어가 1만 개라면, 1개의 1과 9999개의 0이 생기겠죠? 이는 메모리 낭비가 심할 수 있습니다.

## 4. 임베딩 (Embedding): 단어의 지도 만들기
원-핫 인코딩은 단어의 **의미**를 모릅니다. (사과와 배가 비슷한지 모름)
그래서 등장한 것이 **임베딩**입니다.

**비유**:
- 단어들을 넓은 운동장에 풀어놓습니다.
- 뜻이 비슷한 단어끼리는 가까이, 다른 단어는 멀리 배치합니다.
- 각 단어의 **위치 좌표(GPS)**가 바로 임베딩 벡터입니다.
    - 사과: [1.2, 3.5]
    - 배: [1.3, 3.4] (사과랑 가까움!)
    - 자동차: [9.9, -5.2] (사과랑 멂!)

(이 부분은 나중에 Word2Vec 실습에서 더 자세히 다룹니다!)
