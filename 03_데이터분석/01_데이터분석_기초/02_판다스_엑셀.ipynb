```python
# 판다스 라이브러리 임포트
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

# 데이터프레임 생성
df = pd.DataFrame(data)                           # 데이터프레임(표) 생성

# 단일 열 선택
print(df['Name'])
# 인덱스를 기반으로 행 선택
# 조건/라벨 기반 데이터 선택 및 수정
print(df.loc[1])  # 두 번째 행 (인덱스 0부터 시작)
# 위치(인덱스) 기반 데이터 선택
print(df.iloc[2])  # 세 번째 행 (순서 1부터 시작)
```

```python
# 나이가 30 이상인 행 선택
# 조건에 맞는 데이터 필터링
print(df[df['Age'] >= 30])
# 조건에 맞는 데이터 필터링
print(df['Age'] >= 30) # 조건만 확인(True/False)
```

```python
df['contry'] = ['USA', 'Canada', 'Mexico', 'Korea']
print(df)
```

```python
# 새로운 행 추가
# 조건/라벨 기반 데이터 선택 및 수정
df.loc[4] = ['Eve', 28, 'San Francisco', 'USA']   # 이름으로 데이터 선택
print(df)
```

```python
# contry 열 제거
# 불필요한 행/열 삭제
df = df.drop('contry', axis=1)                    # 삭제하기
print(df)
```

```python
# 0번인덱스 행 삭제
# 불필요한 행/열 삭제
df = df.drop(0)                                   # 삭제하기
print(df)
```

```python
# inplace => 원본 데이터프레임 자체를 변경
# 불필요한 행/열 삭제
df = df.drop(2, inplace=True)                     # 삭제하기
df
```

```python
# data프레임에서 약간의 공식이 있다.
# df.loc[행, 열] = 값
# 조건/라벨 기반 데이터 선택 및 수정
df.loc[1, 'Age'] = 31                             # 이름으로 데이터 선택
df
```

```python
# 데이터 타입 변환
df['Age'] = df['Age'].astype(float)
print(df.dtypes)
```

```python
# 조건에 맞는 데이터 필터링
df_filtered = df[df['Age'] >= 30]
print(df_filtered)
```

```python
# 조건에 맞는 데이터 필터링
df_filtered = df[(df['Age'] >= 30) & (df['City'] == 'Chicago')]
print(df_filtered)
```

### **[이론] 데이터 변환 및 병합**

```python
# 판다스 라이브러리 임포트
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 90, 78, 92]
}

# 데이터프레임 생성
df = pd.DataFrame(data)                           # 데이터프레임(표) 생성

# apply()를 이용한 데이터 변환
# 함수 적용 (데이터 변환)
df['Adjutsted_score'] = df['Score'].apply(lambda x: x + 10)
print(df)
```

```python

# 단순 연산
df['adjustedScore'] = df['Score'] + 10
print(df)
```

```python
# map()을 이용한 값 매핑
# 시리즈 생성
gender_data = pd.Series(['Male', 'Female', 'Female', 'Male'])  # 시리즈(한 줄 데이터) 생성

# 값 매핑 (Mapping)
gender_numeric = gender_data.map({'Male': 0, 'Female': 1})
print(gender_numeric)
```

```python
# merge(): 두 데이터프레임 병합
# 데이터프레임 생성
df1 = pd.DataFrame(                               # 데이터프레임(표) 생성
    {
        'CustomerID': [1, 2, 3], 
        'Product': ['Laptop', 'Mouse', 'Keyboard']
    }
)
# 데이터프레임 생성
df2 = pd.DataFrame(                               # 데이터프레임(표) 생성
    {
        'CustomerID': [2, 3, 4], 
        'Location': ['USA', 'Canada', 'Mexico']
    }
)

# 데이터 병합 (Merge)
df_merged = pd.merge(df1, df2, on="CustomerID", how="left")
print(df_merged)
```

```python
# concat(): 데이터 연결
# 데이터프레임 생성
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']})  # 데이터프레임(표) 생성
# 데이터프레임 생성
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'B': ['B2', 'B3']})  # 데이터프레임(표) 생성

# 데이터 연결 (Concat)
df_concat = pd.concat([df1, df2], axis=0, ignore_index=True)  # 인덱스 확인
print(df_concat)
```

```python
# join(): 인덱스 기준 병합
# 데이터프레임 생성
df1 = pd.DataFrame({'A': ['A0', 'A1']}, index=['K0', 'K1'])  # 데이터프레임(표) 생성
# 데이터프레임 생성
df2 = pd.DataFrame({'B': ['B0', 'B1']}, index=['K0', 'K1'])  # 데이터프레임(표) 생성

df_join = df1.join(df2)
print(df_join)
```

```python
# pivot(): 데이터 재구조화
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02'],
    'City': ['New York', 'Los Angeles', 'New York'],
    'Temperature': [55, 68, 52]
    }

# 데이터프레임 생성
df_weather = pd.DataFrame(data)                   # 데이터프레임(표) 생성
# 피벗 테이블 생성
df_pivot = df_weather.pivot(index='Date', columns='City', values='Temperature')  # 컬럼(열) 이름 확인
print(df_pivot)
```

```python
# pivot_table(): 집계 함수 적용 가능한 피벗
sales_data = {
    'Product': ['A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 250],
    'Region': ['North', 'South', 'North', 'South']
}

# 데이터프레임 생성
df_sales = pd.DataFrame(sales_data)               # 데이터프레임(표) 생성
# 피벗 테이블 생성 (집계 포함)
df_pivot_table = df_sales.pivot_table(values='Sales', index='Region', columns='Product', aggfunc='sum')
print(df_pivot_table)
```

```python
# melt(): 넓은 형식 -> 긴 형식
# 데이터프레임 생성
df_wide = pd.DataFrame({                          # 데이터프레임(표) 생성
    'ID': [1, 2, 3],
    'Math': [85, 90, 88],
    'English': [78, 85, 90]
})

# 데이터 구조 변환 (Wide -> Long)
df_melt = pd.melt(df_wide, id_vars=['ID'], value_vars=['Math', 'English'], var_name='Subject', value_name='Score')
print(df_melt)
```

```python
# groupby(): 그룹화 연산
data = {
    'Department': ['HR', 'IT', 'HR', 'Finance', 'IT', 'Finance'],
    'Salary': [50000, 60000, 52000, 80000, 65000, 90000]
}
# 데이터프레임 생성
df = pd.DataFrame(data)                           # 데이터프레임(표) 생성

# 데이터 그룹화 및 집계
grouped = df.groupby('Department')['Salary'].mean()  # 평균 계산
print(grouped)
```

```python
# 판다스 라이브러리 임포트
import pandas as pd
import matplotlib.pyplot as plt

# 예시 데이터
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
        'Sales': [200, 300, 400, 500, 450]}
# 데이터프레임 생성
df = pd.DataFrame(data)                           # 데이터프레임(표) 생성

# 선 그래프 그리기
df.plot(x='Month', y='Sales', kind='line', title='Monthly Sales')
plt.show()                                        # 그래프 보여주기
```

```python
# 막대 그래프 그리기
df.plot(x='Month', y='Sales', kind='bar', title='Monthly Sales')
plt.show()                                        # 그래프 보여주기
```

```python
# 히스토그램 그리기
df['Sales'].plot(kind='hist', title='Sales Distribution')
plt.show()                                        # 그래프 보여주기
```

```python
# 날짜 데이터를 datetime 형식으로 변환
df['Date'] = pd.to_datetime(df['Month'] + ' 2024')
print(df)
```

```python
df.set_index('Date', inplace=True)                # 인덱스 확인
monthly_sales = df.resample('M').sum()            # 합계 계산
monthly_sales
```

```python
# 시계열 데이터 시각화
monthly_sales.plot(title='Monthly Sales Over Time')
plt.show()                                        # 그래프 보여주기
```

```python
df['RollingMean'] = df['Sales'].rolling(window=3).mean()  # 평균 계산
df[['Sales', 'RollingMean']].plot(title='Sales and 3-Month Rolling Average')
plt.show()                                        # 그래프 보여주기
# df[['Sales', 'RollingMean']]
```
