```python
from huggingface_hub import create_repo, upload_folder
from dotenv import load_dotenv
import os

load_dotenv()
HF_TOKEN = os.environ["HF_TOKEN"]

# 업로드

repo_id = "tree0327/vclm-korean-7b-orca-math-korean-dpo"
create_repo(
    repo_id=repo_id, repo_type="model", exist_ok=True
)  # 레포 없으면 생성, 있으면 넘어감

upload_folder(
    repo_id=repo_id,
    repo_type="model",
    folder_path="vclm-korean-7b-orca-math-korean-dpo",
    commit_message="Copy model from capybaraOh",
)
print("업로드 완료:", repo_id)
```
