## 3. ì‹¤ìŠµ: í•œêµ­ì–´ ì „ì²˜ë¦¬ (KoNLPy)
íŒŒì´ì¬ì—ëŠ” **KoNLPy(ì½”ì—”ì—˜íŒŒì´)**ë¼ëŠ” í›Œë¥­í•œ í•œêµ­ì–´ ì²˜ë¦¬ ë„êµ¬ê°€ ìˆìŠµë‹ˆë‹¤. ê·¸ ì¤‘ `Okt(Open Korean Text)` ë¶„ì„ê¸°ë¥¼ ì¨ë´…ì‹œë‹¤.

```python
# konlpy ì„¤ì¹˜ í•„ìš” (!pip install konlpy)
# ì„¤ì¹˜ê°€ ì•ˆ ë˜ì–´ ìˆë‹¤ë©´ ì´ ì…€ì˜ ì£¼ì„ì„ í’€ê³  ì‹¤í–‰í•˜ì„¸ìš”.
# !pip install konlpy

from konlpy.tag import Okt

# 1. ë„êµ¬ ê°€ì ¸ì˜¤ê¸° (ì¹¼ ì¤€ë¹„)
okt = Okt()

text = "ì™€! ì§„ì§œ ë§›ìˆëŠ” ì‚¬ê³¼ë¥¼ ë¨¹ì—ˆë‹¤. ì•„, ë°°ë¶ˆëŸ¬."

# 2. ëª…ì‚¬ë§Œ ì¶”ì¶œí•˜ê¸° (ì•Œì§œë°°ê¸° ì¬ë£Œ)
nouns = okt.nouns(text)
print("ëª…ì‚¬ ì¶”ì¶œ:", nouns)

# 3. í˜•íƒœì†Œ ë¶„ì„ (ì¬ë£Œ ì°ê¸°)
morphs = okt.morphs(text)
print("í˜•íƒœì†Œ ë¶„ì„:", morphs)

# 4. í’ˆì‚¬ íƒœê¹… (ì¬ë£Œ ì´ë¦„í‘œ ë¶™ì´ê¸°)
pos = okt.pos(text)
print("í’ˆì‚¬ íƒœê¹…:", pos)
```
