```python
# set 자료형은 비시퀀스이기 떄문에 순서가 없다.
# 중괄호로 작성하는 것은 dictionary와 같다.
# 하지만 dictionary는 key:value의 쌍으로 이루워진데 반해
# set는 key(value) 하나로만 구성한다.
# 중복되는 요소값을 갖지 않는다.
```

```python
set_var = {"key1", "key2" }
```

```python
print(type(set_var))
```

```python
print(set_var)
```

```python
# 빈 세트는 set() 함수로만 작성할 수 있다.
var1 = {}
```

```python
print(type(var1)) # 빈 {}는 dictionary가 생성된다
```

```python
var2 = set()
```

```python
print(type(var2))
```

```python
print(var2)
```

```python
# set 자료형에 요소를 추가할 때는 add() 메소드를 사용한다.
# 이미 갖고 있는 요소를 추가하면 추가되지 않는다.

var2.add(1)
```

```python
print(var2)
```

```python
var2.add(1) # 이미 갖고 있는 요소를 한번 더 추가
```

```python
print(var2) # 추가되지 않는 것을 확인할 수 있다.
```

```python
# set 자료형의 요소를 삭제할 때는 remove 메소드를 호출한다.
# 없는 요소를 삭제하려고 하면 에러가 난다.
# discard() 메소드로 없는 요소를 삭제하려고 해도 에러가 나지 않는다.
var2.remove(1)
```

```python
print(var2)
```

```python
# set 자료형의 합집합(union)
a = {1, 2, 3, 4}
b = {2, 4, 5}
```

```python
print(a)
```

```python
print(set.union(a, b))
```

```python
print(a | b)
```

```python
# set 자료형의 교집합(intersection)
print(set.intersection(a, b))
```

```python
print(a & b)
```

```python
# set 자료형의 차집합(difference)
print(set.difference(a, b))
```

```python
print(a - b)
```
