```python
# List의 값 변경하기
# 인덱싱을 통해서 하나의 요소의 값을 변경
cubes = [1, 8, 27, 65, 125] # 65가 잘못된 갑이라서 64로 변경해야함
cubes[3] = 4**3 # 인덱싱을 통한 값 변경
print(cubes)
```

```python
# Slicing을 통해서 여러개의 요소 변경하기
letters = ["c", "d", "e"] # ["a", "b", "e"]
letters[0:2] = ["a", "b"] # 슬라이싱을 통한 요소들의 변경
                          # 요소의 개수(len)가 같아야 한다.
print(letters)
```

```python
#리스트 요소의 개수 구하기
print(len(letters))
```

```python
# List 요소에 포함 여부를 확인하는 방법으로 in 이라는 키워드를 사용
numbers = [1, 2, 3, 4, 5, 6, 7]
```

```python
print(3 in numbers)
```

```python
print(-3 in numbers)
```

```python
# 리스트변수.append(추가하고 싶은 요소)
print(numbers)
```

```python
numbers.append(8) # 8이라는 요소를 맨 뒤에 삽입
```

```python
print(numbers)
```

```python
# 리스트변수.insert(인덱스, 추가하고 싶은 요소)
numbers.insert(0, 0) #0번째 인덱스에 0 추가
```

```python
print(numbers)
```

```python
numbers.insert(4, 0)
```

```python
print(numbers)
```

```python
numbers.insert(6, 0)
```

```python
print(numbers)
```

```python
numbers.insert(len(numbers), 0)
```

```python
print(numbers)
```

```python
# 리스트변수.extend(리스트변수2)
# 리스트변수1에 제일 뒤쪽으로 리스트변수2의 요소들을 순서태로 추가한다.
numbers.extend([9, 10, 11])
```

```python
print(numbers)
```

```python
# 리스트변수.remove(삭제하고 싶은 요소값)
# 삭제하고 싶은 요소의 값과 동일한 첫번째 요소를 제거합니다.
print(numbers)
```

```python
numbers.remove(0) #인덱스 순서로 일치하는 요소를 찾아 제거
```

```python
print(numbers)
```

```python
# 리스트변수.pop() 제일 마지막 요소를 삭제한 후 반환
print(numbers.pop())
```

```python
print(numbers)
```

```python
print(numbers.append(11))
```

```python
print(numbers)
```

```python
# 리스트변수.index(인덱스를 찾고자하는 요소의 값)
# 인덱스 위치를 알고 싶은 요소의 값과 일치하는 인덱스 순서로 봤을 때 첫번째로 등장한 요소의 인덱스 번호를 반환
print(numbers)
```

```python
numbers.index(0)
```

```python
# 리스트변수.count(요소의 값)
# 소괄호 안에 넣어준 요소의 값(인자)와 같은 것의 개수를 반환
print(numbers.count(0))
```

```python
# 리스트변수.sort()
# 리스트변수를 오름차순으로 정렬
numbers.sort()
```

```python
print(numbers)
```

```python
# 리스트변수.revers()
# 리스트변수의 요소를 반대 방향으로 뒤집는다.
numbers.reverse()
```

```python
print(numbers)
```

```python
# list로 stack 구조 활용하기
# 3, 4, 5라는 값을 순서대로 넣기
# 6이라는 값과 7이라는 값을 차례대로 넣기
# 제일 바깥쪽에 있는(Last)에 있는 값을 순서대로 3개 꺼내기
stack_var = [3, 4, 5]
stack_var_sub = [6, 7]
stack_var.extend(stack_var_sub)
```

```python
print(stack_var)
```

```python
print(stack_var.pop())
```

```python
print(stack_var.pop())
```

```python
print(stack_var.pop())
```

```python
print(stack_var)
```
