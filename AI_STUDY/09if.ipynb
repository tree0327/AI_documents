```python
# if문
# if문은 특정 조건에서만 실행되는 코드블록을 작성할 때 사용한다.
# if선언 후 바로 다음줄에는 무조건 들여쓰기가 나와야 한다.
# 1=1 처럼 대입연산자는 같다는 의미로 사용하지 않는다. 같다는 의미를 표현할떄는 ==를 사용

if True:
    print("hello world") # 조건식이 True일 때 실행된다.
print("파이썬 공부중입니다")
```

```python
if False:
    print("hello world") #조건식이 False일때는 실행되지 않는다.
print("파이썬 공부중입니다")
```

```python
int_value = 10
if type(int_value) is str : # int_value 변수의 type이 int 라면
    print("int 자료형입니다.") # 거짓이면 실행되지 않습니다.
print("이 줄은 무조건 실행됩니다.") # 참과 거짓에 상관없이 실행됩니다.
```

```python
# 복합적인 조건을 따져서 실행하는 if문을 작성하고 싶을 때는 논리연산자를 활용한다.
int_value2 = 100

#int_value2의 값이 10보다 크고, int값을 가질 때에만 "안녕하세요" 출력
if int_value2 > 10 and type(int_value) is int:
    print("안녕하세요")
print("안녕하세요면 True 이거만 있으면 False")
```

```python
# if와 else를 활용한 분기
score = 60

# score가 60점 이상일 때는 합격입니다 출력
# score가 60점 미만일 때는 합격입니다 출력
if score >= 60:
    print("합격입니다.")
else:
    print("불합격입니다.")
```

```python
# if와 elif를 활용한 분기
age = 40

# 20대면 20대입니다 출력
# 30대면 30대입니다 출력
# 40대면 40대입니다 출력

if 20 <= age < 30:
    print("20대입니다.")
elif 30 <= age < 40:
    print("30대입니다.")
elif 40 <= age < 50:
    print("40대입니다.")
```

```python
# 10대, 20대, 30대를 체크하고 40대 이상이면 불혹을 넘기셨군요 출력
age2 = 60
# 20대 미만이면 미성년자입니다 출력
# 20대면 20대입니다 출력
# 30대면 30대입니다 출력
# 40대면 40대입니다 출력
if age2 < 20:
    print("미성년자입니다.")
if 20 <= age2 < 30:
    print("20대입니다.")
elif 30 <= age2 < 40:
    print("30대입니다.")
elif 40 <= age2:
    print("불혹을 넘기셨습니다.")
```

```python
# 성적이 90점 이상이면 A
# 성적이 80점 이상 90점 미만이면 B
# 성적이 70점 이상 80점 미만이면 C
# 성적이 60점 이상 70점 미만이면 D
# 성적이 50점 이상 60점 미만이면 E
# 성적이 50점 미만이면 F를 출력한다.
score2 = 50

print("성적을 알려드립니다.")

if score2 >= 90:
    print("A")
elif 80 <= score2 < 90:
    print("B")
elif 70 <= score2 < 80:
    print("C")
elif 60 <= score2 < 70:
    print("D")
elif 50 <= score2 < 60:
    print("E")
else:
    print("F")
```
