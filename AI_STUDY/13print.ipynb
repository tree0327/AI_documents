```python
# 화면에 출력
# print() 내장함수를 활용하여 할 수 있다.
print("Hello world")
```

```python
# sep라는 keyword argument가 기본값(default)으로 " "(공백)을 가지고 있다.
print(1, 1.1, [1, "가", 1.3], sep=" ")
```

```python
print(1, 1.1, [1, "가", 1.3], sep="구분")
```

```python
# end라는 keyword argument가 기본값()default)으로 "\n" 개행문자를 가지고 있다.
print("출력할 때 쓰는 print() 내장 함수는" )
print("기본적으로 개행문자를 마지막문자로 사용한다.")
```

```python
print("출력할 때 쓰는 print() 내장 함수는", end=" -마지막 문자를 변경- ")
print("기본적으로 개행문자를 마지막문자로 사용한다.")
```
