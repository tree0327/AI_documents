```python
# dictionary : 비시퀀스 자료형
# 순서가 없기 때문에 인덱스가 없다.
# 시퀀스 자료형은 특정 값에 접근하기 위해 인덱스를 활용
# 비시퀀스 자료형은 dictionary는 key:value로 쌍으로 이루어지기 때문에
# key를 통해서 value에 접근
dic_var = {"key" : "value"}
```

```python
print(type(dic_var))
```

```python
print(dic_var)
```

```python
# key라는 것은 식별을 위한 것이기 때문에
# 중복된 값을 가질 수 없다.
dic_var2 = {"key":"value", "key":"last value"}
```

```python
print(dic_var2)
```

```python
# key로 사용할 수 있는 자료형
# 문자열(str), 정수(int), 실수(float), 불(bool), 튜플(tuple)
# dictionary 안에서 key는 단 하나의 타입만을 사용해야 하는것은 x
# key 타입으로 여러가지를 동시에 가질 수 있다.
dic_var3 = {
    "key" : "value",
    1 : "value2",
    1.1 : "value3",
    True : "value4",
    ("key", 1, 1.1, True) : "value4"
}
```

```python
print(dic_var3)
```

```python
# value는 모든 자료형을 사용할 수 있다.
dic_var4 = {
    "key" : "value",
    2: 2,
    1.1: 1.1,
    True: False,
    ("key", 1, 1.1, True): ("key", 1, 1.1, True),
    "list": [1, 1.1, True],
    "dict": {"key" : "value"},
    "set": {"key" : "key2"}
}
```

```python
print(dic_var4)
```

```python
print(list(map(type, dic_var4.values())))
# dictionary는 value로 모든 type의 값이 올 수 있다.
```

```python
# 빈 dictionary는 {}, dict()라는 arguments가 없는 함수 호출을 통해서 생성할 수 있다.
empty_dict = {}
print(empty_dict)
```

```python
empty_dict2 = dict()
print(empty_dict2)
```

```python
# dictionary는 dict() 함수를 통해 생성할 수 있다.
# 이때 arguments는 표현에 제약이 존재하기 때문에
# key값이 문자열로만 존재하게 된다.
dict_a = dict(
    one = "value1",
    two = "value2"
)
```

```python
print(type(dict_a))
```

```python
print(dict_a)
```

```python
# dict() 함수를 사용해서 key의 type을 str(문자열)이 아닌 것으로 할당할 수 있다.
dict_4 = dict(
    [
        ["key", "value"],
        [1, "value2"]
    ]
)
```

```python
print(dict_4)
```

```python
# key의 값들을 갖는 리스트
# value의 값들을 갖는 리스트가 각각 별개로 존재할 때
# zip() 함수에 이 각각의 리스트들을 arguments로 넣어주면
# 인덱스가 같은 요소들을 꺼내서 하나의 쌍으로 튜플로 만들어줍니다.
# dict() 함수와 함꼐 사용하면 key의 리스트와 value의 리스트를
# dictionary 자료형으로 만들 수 있다.

keys = [
    "key1", "key2", "key3"
]
values = [
    "value1", "value2", "value3"
]

print(dict(zip(keys, values)))
```

```python
dict_d = dict(zip(keys, values))
```

```python
print(dict_d)
```

```python
print(dict_d["key2"])
```

```python
dict_d["key1"] = "super value"
```

```python
print(dict_d["key1"])
```

```python
# 없는 key값 = value를 지정하면 새로운 key, value가 생성된다
dict_d["key4"] = "value4" #dict_d에 key4라는 key는 현재 없음.
```

```python
print(dict_d) #key4, value4가 추가된 걸 확인할 수 있다.
```

```python
# dictionary 자료형에서 특정 key값이 존재하는지 확인할 때 in 을 사용한다.
print("key1" in dict_d)
```

```python
# dictionary 자료형에서 특정 value값이 존재하는지 확인할 때
# dictionary values를 구해야한다.
print(dict_d.values())
```

```python
print("super value" in dict_d.values())
```

```python
print(dict_d)
```

```python
del dict_d["key4"]
```

```python
print(dict_d)
```

```python
# dictionary 자료형에서 element(요소)의 개수는 len()함수를 통해 구한다.
# 이때 요소의 개수:값의 개수 = 1:1로 매핑된다.
print(len(dict_d))
```
