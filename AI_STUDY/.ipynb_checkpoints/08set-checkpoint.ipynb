```python
# set 자료형은 비시퀀스이기 떄문에 순서가 없다.
# 중괄호로 작성하는 것은 dictionary와 같다.
# 하지만 dictionary는 key:value의 쌍으로 이루워진데 반해
# set는 key(value) 하나로만 구성한다.
# 중복되는 요소값을 갖지 않는다.
```

```python
set_var = {"key1", "key2"}
```

```python
print(type(set_var))
```

```python
print(set_var)
```

```python
# 빈 세트는 set() 함수로만 작성할 수 있다.
var1 = {}
```

```python
print(type(var1)) # 빈 {}는 dictionary가 생성된다
```

```python
var2 = set()
```

```python
print(type(var2))
```

```python
print(var2)
```
