```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.models import Sequential

# 3. CNN ëª¨ë¸ ì¡°ë¦½í•˜ê¸°
model = Sequential()

# (1) ëˆˆ ë‹¬ê¸°: í•„í„° 32ê°œë¡œ í›‘ì–´ë³´ê¸° (Conv2D)
# kernel_size=(3,3): 3x3 í¬ê¸°ì˜ ë‹ë³´ê¸°
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))

# (2) ìš”ì•½í•˜ê¸°: í¬ê¸° ì ˆë°˜ìœ¼ë¡œ ì¤„ì´ê¸° (MaxPooling)
model.add(MaxPooling2D(pool_size=(2, 2)))

# (3) íŒë‹¨í•˜ê¸°: í´ì„œ ê²°ê³¼ ë‚´ê¸° (Flatten -> Dense)
model.add(Flatten())
model.add(Dense(10, activation='softmax')) # 0~9 ìˆ«ì ë¶„ë¥˜

model.summary()
```
