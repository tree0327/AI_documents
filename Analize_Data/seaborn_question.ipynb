```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

sns.countplot(x='survived', data=titanic, palette='muted')

plt.title('ÏÉùÏ°¥Ïûê Î∞è ÏÇ¨ÎßùÏûê Ïàò')
plt.xticks(ticks=[0, 1], labels=['ÏÇ¨Îßù', 'ÏÉùÏ°¥'])


plt.show()                                        # Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞
```

```python
titanic
sns.barplot(x='sex', y='survived', data=titanic, palette='pastel')
plt.title('ÏÑ±Î≥ÑÏóê Îî∞Î•∏ ÏÉùÏ°¥Î•†')
plt.xticks(ticks=['male', 'female'], labels=['ÎÇ®Ïûê', 'Ïó¨Ïûê'])
plt.show()                                        # Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞
```

```python
sns.barplot(x='class', y='survived', hue='sex', data=titanic, palette='pastel')
plt.title('ÌÅ¥ÎûòÏä§ÏôÄ ÏÑ±Î≥ÑÏóê Îî∞Î•∏ ÏÉùÏ°¥Î•†')
plt.xticks(ticks=['First', 'Second', 'Third'], labels=['1Îì±ÏÑù', '2Îì±ÏÑù', '3Îì±ÏÑù'])
plt.show()                                        # Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞
```

```python
sns.histplot(titanic['age'].dropna(), kde=True, bins=30, color='purple')  # ÏÇ≠Ï†úÌïòÍ∏∞
plt.title('ÎÇòÏù¥ Î∂ÑÌè¨')
plt.xlabel('ÎÇòÏù¥')
plt.ylabel('ÎπàÎèÑÏàò')
plt.show()                                        # Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞
```
