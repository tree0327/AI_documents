# üìâ MatplotlibÏúºÎ°ú Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
ÌååÏù¥Ïç¨ÏóêÏÑú Í∞ÄÏû• Í∏∞Î≥∏Ï†ÅÏù¥Í≥† Ïú†Î™ÖÌïú ÏãúÍ∞ÅÌôî ÎùºÏù¥Î∏åÎü¨Î¶¨Ïù∏ `matplotlib`ÏùÑ Î∞∞ÏõåÎ¥ÖÏãúÎã§.
ÏÑ† Í∑∏ÎûòÌîÑ, ÎßâÎåÄ Í∑∏ÎûòÌîÑ, ÏÇ∞Ï†êÎèÑ Îì± Îã§ÏñëÌïú Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¥ Ïàò ÏûàÏñ¥Ïöî.


## 1. Í∏∞Î≥∏ ÏÑ† Í∑∏ÎûòÌîÑ (Line Plot)
Í∞ÄÏû• Í∞ÑÎã®Ìïú ÏÑ† Í∑∏ÎûòÌîÑÎ•º Í∑∏Î†§Î¥êÏöî. `plt.plot()` Ìï®ÏàòÏóê Ïà´ÏûêÎì§ÏùÑ ÎÑ£Ïñ¥Ï£ºÎ©¥ Îê©ÎãàÎã§.


```python
import matplotlib.pyplot as plt

# Í∞ÑÎã®Ìïú ÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.plot(
    [1, 2, 3, 4],
    [10, 20, 25, 30]
)

plt.title("Simple Line Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

## 2. Ïó¨Îü¨ Í∑∏ÎûòÌîÑ ÎèôÏãúÏóê Í∑∏Î¶¨Í∏∞ (Subplots)
`subplots`Î•º ÏÇ¨Ïö©ÌïòÎ©¥ ÎèÑÌôîÏßÄ(Figure)Î•º Ïó¨Îü¨ Ïπ∏ÏúºÎ°ú ÎÇòÎàÑÏñ¥ Ïó¨Îü¨ Í∑∏ÎûòÌîÑÎ•º Ìïú Î≤àÏóê Î≥¥Ïó¨Ï§Ñ Ïàò ÏûàÏñ¥Ïöî.
- Ìñâ(row): Í∞ÄÎ°ú Ï§Ñ Í∞úÏàò
- Ïó¥(column): ÏÑ∏Î°ú Ï§Ñ Í∞úÏàò


```python
# Ïó¨Îü¨Í∞±Ïùò ÏÑúÎ∏åÌîåÎ°Ø ÏÉùÏÑ±
fig, axs = plt.subplots(2, 2)

# Í∞Å ÏÑúÎ∏åÌîåÎ°ØÏóê Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
axs[0, 0].plot([1, 2, 3], [1, 2, 3])
axs[0, 1].plot([1, 2, 3], [1, 4, 9])
axs[1, 0].plot([1, 2, 3], [1, 8, 27])
axs[1, 1].plot([1, 2, 3], [1, 16, 81])

plt.show()
```

## 3. Í∞ùÏ≤¥ ÏßÄÌñ• Î∞©Ïãù (Figure & Axes)
Í∑∏ÎûòÌîÑÎ•º Ï¢Ä Îçî ÏÑ∏Î∞ÄÌïòÍ≤å Îã§Î£®Í∏∞ ÏúÑÌï¥ **Í∞ùÏ≤¥(Object)**Î•º ÏÇ¨Ïö©ÌïòÎäî Î∞©Î≤ïÏù¥ÏóêÏöî.
- `Figure`: Ï†ÑÏ≤¥ ÎèÑÌôîÏßÄ
- `Axes`: Í∑∏ÎûòÌîÑÍ∞Ä Í∑∏Î†§ÏßÄÎäî Í∞ÅÍ∞ÅÏùò Ïπ∏


```python
import matplotlib.pyplot as plt

# FigureÏôÄ Axes Í∞ùÏ≤¥ ÏÉùÏÑ±
fig, ax = plt.subplots()

# Axes Í∞ùÏ≤¥ÏóêÏÑú Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Object-oriented Graph')

plt.show()
```

### üìä ÏòàÏ†ú Îç∞Ïù¥ÌÑ∞Î°ú Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
ÏöîÏùºÎ≥Ñ Îç∞Ïù¥ÌÑ∞ ÏòàÏ†úÎ•º ÎßåÎì§Ïñ¥Î¥ÖÏãúÎã§.


```python
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [22, 24, 19, 23, 25, 20, 18]

# ÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.plot(days, temperatures, marker='o', linestyle='-', color='b')
plt.title('Temperature Variation Over a Week')
plt.xlabel('Day')
plt.ylabel('Temperature (¬∞C)')
plt.legend(['Temperature'])
plt.grid(True)
plt.show()
```

### üåå ÏÇ∞Ï†êÎèÑ (Scatter Plot) Ïó∞Ïäµ
Îëê Î≥ÄÏàò Í∞ÑÏùò Í¥ÄÍ≥ÑÎ•º Ï†êÏúºÎ°ú ÌëúÌòÑÌï¥Î¥ÖÎãàÎã§.


```python
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 20, 25, 30]
y2 = [5, 10, 15, 20, 25]

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.plot(x, y1, label='Dataset 1', marker='o')
plt.plot(x, y2, label='Dataset 2', marker='s')

plt.title('Customized Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True)
plt.show()
```

### üç∞ ÏõêÌòï Í∑∏ÎûòÌîÑ (Pie Chart) ÎòêÎäî ÎßâÎåÄ Í∑∏ÎûòÌîÑ
ÎπÑÏú®Ïù¥ÎÇò ÌÅ¨Í∏∞Î•º ÎπÑÍµêÌï† Îïå Ï¢ãÏïÑÏöî.


```python
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [150, 200, 180, 220, 170]

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.bar(months, sales, color='skyblue')
# barh -> ÏàòÌèâÎßâÎåÄÍ∑∏ÎûòÌîÑ 

for i, value in enumerate(sales):
    plt.text(i, value, value, ha='center', va='bottom')

plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.show()
```

### üì¶ ÏÉÅÏûê Í∑∏Î¶º (Box Plot)
Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨ÏôÄ Ïù¥ÏÉÅÏπò(ÌäÄÎäî Í∞í)Î•º ÌôïÏù∏Ìï¥Î¥ÖÎãàÎã§.


```python
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
classA = [85, 88, 90, 93, 95, 97, 99]
classB = [78, 80, 82, 85, 87, 89, 91]
classC = [70, 72, 75, 77, 79, 81, 83]

# Î∞ïÏä§ÌîåÎ°Ø Í∑∏Î¶¨Í∏∞
data = [classA, classB, classC]
plt.boxplot(data, labels=['Class A', 'Class B', 'Class C'])
plt.title('Exam Score Distribution by Class')
plt.ylabel('Score')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
heights = np.random.normal(170, 5, 200)  # ÌèâÍ∑† 170cm, ÌëúÏ§ÄÌé∏Ï∞® 5cmÏù∏ Îç∞Ïù¥ÌÑ∞ 200Í∞ú ÏÉùÏÑ±

# ÌûàÏä§ÌÜ†Í∑∏Îû® Í∑∏Î¶¨Í∏∞
plt.hist(heights, bins=10, color='green', edgecolor='black')
plt.title('Height Distribution of Students')
plt.xlabel('Height (cm)')
plt.ylabel('Number of Students')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
np.random.seed(0)
study_hours = np.random.normal(5, 1.5, 100)  # ÌèâÍ∑† 5ÏãúÍ∞Ñ, ÌëúÏ§ÄÌé∏Ï∞® 1.5ÏãúÍ∞Ñ
scores = study_hours * 10 + np.random.normal(0, 5, 100)  # Í≥µÎ∂Ä ÏãúÍ∞ÑÏóê Îî∞Î•∏ Ï†êÏàò

# ÏÇ∞Ï†êÎèÑ Í∑∏Î¶¨Í∏∞
plt.scatter(study_hours, scores, color='purple', alpha=0.6)
plt.title('Relationship between Study Hours and Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
x = [1, 2, 3, 4, 5]
y = [10, 40, 15, 35, 20]
y2 = [30, 25, 20, 15, 10]
y3 = [10, 15, 18, 24, 30]

# Îã§ÏñëÌïú ÏÑ† Ïä§ÌÉÄÏùº
plt.plot(x, y, label='Solid Line', color='blue', linestyle='-', linewidth=2, marker='o')
plt.plot(x, y2, label='Dash Line', color='green', linestyle='--', linewidth=2, marker='s')
plt.plot(x, y3, label='Dot Line', color='red', linestyle=':', linewidth=2, marker='^')

# Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Line styles')
plt.legend()
plt.grid(True)

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.title('Sine Wave')

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.title('Cosine Wave')

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 2D Î∞∞Ïó¥ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
data = np.random.rand(10, 10)

# Îëê Í∞úÏùò ÏÑúÎ∏åÌîåÎ°ØÏóêÏÑú Îã§Î•∏ Ïª¨Îü¨Îßµ Ï†ÅÏö©
plt.subplot(1, 2, 1)
plt.imshow(data, cmap='plasma')
plt.title('Plasma Colormap')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(data, cmap='inferno')
plt.title('Inferno Colormap')
plt.colorbar()

# ÏÑúÎ∏åÌîåÎ°Ø Í∞ÑÍ≤© Ï°∞Ï†ï
plt.tight_layout()

# Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0) # Ïû¨Ïó∞ÏÑ± ÌôïÎ≥¥

# Îç∞Ïù¥ÌÑ∞
x = np.arange(1, 11)
y = np.random.randint(10, 100, 10)

plt.figure(figsize=(12, 10))

# ÏÑ†Í∑∏ÎûòÌîÑ
plt.subplot(2, 2, 1)
plt.plot(x, y, color='pink', marker='o')
plt.title("Line")

# ÎßâÎåÄÍ∑∏ÎûòÌîÑ
plt.subplot(2, 2, 2)
plt.bar(x, y, color='red' , width=0.5)
plt.title("Bar")

# ÌûàÏä§ÌÜ†Í∑∏Îû®
plt.subplot(2, 2, 3)
plt.hist(y, color='green')
plt.title("Histogram")

# ÏÇ∞Ï†êÎèÑ
plt.subplot(2, 2, 4)
plt.scatter(x, y, color='blue', marker='o')
plt.title("Scatter")

# ÏÇ¨Ïù¥Ï¶à figure ÏÇ¨Ïù¥Ï¶à ÎßûÏ∂îÍ≥†
plt.tight_layout()
plt.show()
```
