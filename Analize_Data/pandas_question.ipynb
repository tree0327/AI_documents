````python
# íŒë‹¤ìŠ¤ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
import pandas as pd

product_data = {
    'ProductID': [1, 2, 3, 1, 4, 2],
    'ProductName': ['Laptop', 'Mouse', 'Monitor', 'Laptop', 'Keyboard', 'Mouse'],
    'Price': [1200, 25, 300, 1150, 45, 22],
    'QuantitySold': [10, 100, 50, 8, 70, 120]
}

# ë°ì´í„°í”„ë ˆì„ ìƒì„±
df = pd.DataFrame(product_data)                   # ë°ì´í„°í”„ë ˆì„(í‘œ) ìƒì„±

# 1. ê°€ê²©(Price)ì„ ê¸°ì¤€ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬
# (ì´ë ‡ê²Œ í•˜ë©´ ê°™ì€ ProductID ì¤‘ì—ì„œ ê°€ê²©ì´ ë‚®ì€ í•­ëª©ì´ ë¦¬ìŠ¤íŠ¸ì˜ ìœ„ìª½ìœ¼ë¡œ ì˜¤ê²Œ ë©ë‹ˆë‹¤)
# ë°ì´í„° ì •ë ¬
df_sorted = df.sort_values(by='Price', ascending=True)  # ê°’ ì •ë ¬í•˜ê¸°

# 2. ProductIDë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì¤‘ë³µ ì œê±°
# keep='first' ì˜µì…˜ì€ ì¤‘ë³µëœ ê°’ ì¤‘ ê°€ì¥ ìœ„ì— ìˆëŠ” ì²« ë²ˆì§¸ ê°’(ì•„ê¹Œ ì •ë ¬í•œ ìµœì €ê°€ í•­ëª©)ì„ ë‚¨ê¹ë‹ˆë‹¤.
# ì¤‘ë³µê°’ ì œê±°
df_result = df_sorted.drop_duplicates(subset='ProductID', keep='first')  # ì‚­ì œí•˜ê¸°

# 3. ì¸ë±ìŠ¤ ì¬ì„¤ì • (ê¹”ë”í•œ ì¶œë ¥ì„ ìœ„í•´ ê¸°ì¡´ ì¸ë±ìŠ¤ë¥¼ ë²„ë¦¬ê³  ìƒˆë¡œ ë§¤ê¹ë‹ˆë‹¤)
# ì¸ë±ìŠ¤ ì´ˆê¸°í™”
df_result = df_result.reset_index(drop=True)      # ì¸ë±ìŠ¤ í™•ì¸

# ê²°ê³¼ í™•ì¸
print(df_result)
````

1. ê° ê³ ê°(`CustomerID`)ë³„ë¡œ ì´ êµ¬ë§¤ ê¸ˆì•¡ì„ ê³„ì‚°í•˜ì„¸ìš”.
2. `Laptop`ì„ êµ¬ë§¤í•œ ê³ ê°ë“¤ì˜ í‰ê·  êµ¬ë§¤ ê¸ˆì•¡ì„ ê³„ì‚°í•˜ì„¸ìš”.
3. ê°€ì¥ ë§ì´ íŒë§¤ëœ ìƒí’ˆ(ìƒí’ˆ ì´ë¦„ê³¼ íŒë§¤ëœ ì´ ìˆ˜ëŸ‰)ì„ êµ¬í•˜ì„¸ìš”.
4. `OrderDate`ê°€ 2024ë…„ 1ì›” 10ì¼ ì´í›„ì¸ ì£¼ë¬¸ë§Œ í•„í„°ë§í•˜ì—¬ ì¶œë ¥í•˜ì„¸ìš”.
5. `Product`ë³„ë¡œ íŒë§¤ëŸ‰ê³¼ ì´ íŒë§¤ ê¸ˆì•¡ì„ ê³„ì‚°í•œ í›„, ì´ë¥¼ `Price` ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ì—¬ ì¶œë ¥í•˜ì„¸ìš”.

````python
# íŒë‹¤ìŠ¤ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
import pandas as pd
# ë°ì´í„°
sales_data = {
    'OrderID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Monitor', 'Mouse', 'Keyboard', 'Laptop', 'Mouse', 'Monitor', 'Keyboard'],
    'Quantity': [1, 2, 1, 1, 2, 1, 1, 3, 1, 1],
    'Price': [1200, 25, 1200, 300, 25, 100, 1200, 25, 300, 100],
    'CustomerID': [1001, 1002, 1001, 1003, 1004, 1002, 1001, 1003, 1005, 1002],
    'OrderDate': ['2024-01-05', '2024-01-06', '2024-01-07', '2024-01-10', '2024-01-12', '2024-01-15', '2024-01-18', '2024-01-20', '2024-01-22', '2024-01-25']
}

# DataFrame ìƒì„±
# ë°ì´í„°í”„ë ˆì„ ìƒì„±
df_sales = pd.DataFrame(sales_data)               # ë°ì´í„°í”„ë ˆì„(í‘œ) ìƒì„±

# 'OrderDate'ë¥¼ ë‚ ì§œ í˜•ì‹ìœ¼ë¡œ ë³€í™˜ (ë¬¸ìì—´ -> datetime)
df_sales['OrderDate'] = pd.to_datetime(df_sales['OrderDate'])
````

````python
# 1. ê° ê³ ê°(CustomerID)ë³„ë¡œ ì´ êµ¬ë§¤ ê¸ˆì•¡ì„ ê³„ì‚°í•˜ì„¸ìš”.
df_sales['TotalAmount'] = df_sales['Price'] * df_sales['Quantity']
# ë°ì´í„° ê·¸ë£¹í™” ë° ì§‘ê³„
customer_total = df_sales.groupby('CustomerID')['TotalAmount'].sum()  # í•©ê³„ ê³„ì‚°

customer_total
````

````python
# ë‹¨ê³„ 1: ê° ì£¼ë¬¸ ê±´ë³„ 'ì´ ê¸ˆì•¡(TotalAmount)' ê³„ì‚°
# (ê°œë³„ ë‹¨ê°€ Price * ìˆ˜ëŸ‰ Quantity)ì„ ê³±í•´ì„œ ìƒˆë¡œìš´ ì»¬ëŸ¼ìœ¼ë¡œ ë§Œë“­ë‹ˆë‹¤.
df_sales['TotalAmount'] = df_sales['Price'] * df_sales['Quantity']

# ë‹¨ê³„ 2: ê³ ê°ë³„(CustomerID)ë¡œ ê·¸ë£¹í™”(groupby)í•˜ì—¬ ì´ ê¸ˆì•¡ì˜ í•©(sum) ì§‘ê³„
# ë°ì´í„° ê·¸ë£¹í™” ë° ì§‘ê³„
customer_total = df_sales.groupby('CustomerID')['TotalAmount'].sum()  # í•©ê³„ ê³„ì‚°

# ë‹¨ê³„ 3: ë³´ê¸° ì¢‹ê²Œ ê²°ê³¼ ì¸ë±ìŠ¤ ë¦¬ì…‹ (ì„ íƒ ì‚¬í•­)
# ì¸ë±ìŠ¤ ì´ˆê¸°í™”
customer_total_df = customer_total.reset_index()  # ì¸ë±ìŠ¤ í™•ì¸

print("== ê³ ê°ë³„ ì´ êµ¬ë§¤ ê¸ˆì•¡ ==")
print(customer_total_df)
````

````python
# ------------------------------------------------------------------------------------------------
# 2. `Laptop`ì„ êµ¬ë§¤í•œ ê³ ê°ë“¤ì˜ í‰ê·  êµ¬ë§¤ ê¸ˆì•¡ì„ ê³„ì‚°í•˜ì„¸ìš”.
# ------------------------------------------------------------------------------------------------

# ë‹¨ê³„ 1: 'Laptop'ì„ êµ¬ë§¤í•œ ë°ì´í„°ë§Œ í•„í„°ë§
laptop_sales = df_sales[df_sales['Product'] == 'Laptop']

# ë‹¨ê³„ 2: Laptop êµ¬ë§¤ ê³ ê° ì§‘ë‹¨ì˜ 'TotalAmount' í‰ê·  ê³„ì‚°
laptop_avg_amount = laptop_sales['TotalAmount'].mean()  # í‰ê·  ê³„ì‚°

print(f"\n== Laptop êµ¬ë§¤ ê³ ê° í‰ê·  êµ¬ë§¤ì•¡: {laptop_avg_amount:.2f} ==")
````

````python
# ------------------------------------------------------------------------------------------------
# 3. ê°€ì¥ ë§ì´ íŒë§¤ëœ ìƒí’ˆ(ìƒí’ˆ ì´ë¦„ê³¼ íŒë§¤ëœ ì´ ìˆ˜ëŸ‰)ì„ êµ¬í•˜ì„¸ìš”.
# ------------------------------------------------------------------------------------------------

# ë‹¨ê³„ 1: ìƒí’ˆë³„(Product) íŒë§¤ ìˆ˜ëŸ‰(Quantity) í•©ê³„ ê³„ì‚°
# ë°ì´í„° ê·¸ë£¹í™” ë° ì§‘ê³„
product_quantity = df_sales.groupby('Product')['Quantity'].sum()  # í•©ê³„ ê³„ì‚°

# ë‹¨ê³„ 2: íŒë§¤ ìˆ˜ëŸ‰ì´ ê°€ì¥ ë§ì€(max) ìƒí’ˆ ì°¾ê¸° - idxmax() ì‚¬ìš© (ì¸ë±ìŠ¤ ë°˜í™˜)
# ìµœëŒ€ê°’ì„ ê°€ì§„ ì¸ë±ìŠ¤ ë°˜í™˜
top_product_name = product_quantity.idxmax()
top_product_qty = product_quantity.max()

print(f"\n== ê°€ì¥ ë§ì´ íŒë§¤ëœ ìƒí’ˆ: {top_product_name} ({top_product_qty}ê°œ) ==")
````

````python
# ------------------------------------------------------------------------------------------------
# 4. `OrderDate`ê°€ 2024ë…„ 1ì›” 10ì¼ ì´í›„ì¸ ì£¼ë¬¸ë§Œ í•„í„°ë§í•˜ì—¬ ì¶œë ¥í•˜ì„¸ìš”.
# ------------------------------------------------------------------------------------------------

# ë‹¨ê³„ 1: ë‚ ì§œ ì¡°ê±´ í•„í„°ë§ (> '2024-01-10')
recent_orders = df_sales[df_sales['OrderDate'] > '2024-01-10']

print("\n== 2024ë…„ 1ì›” 10ì¼ ì´í›„ ì£¼ë¬¸ ==")
print(recent_orders)
````

````python
# ------------------------------------------------------------------------------------------------
# 5. `Product`ë³„ë¡œ íŒë§¤ëŸ‰ê³¼ ì´ íŒë§¤ ê¸ˆì•¡ì„ ê³„ì‚°í•œ í›„, ì´ë¥¼ `Price` ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ì—¬ ì¶œë ¥í•˜ì„¸ìš”.
# ------------------------------------------------------------------------------------------------

# ë‹¨ê³„ 1: ìƒí’ˆë³„(Product)ë¡œ ê·¸ë£¹í™”í•˜ì—¬ -> ìˆ˜ëŸ‰(Quantity)ê³¼ ì´ê¸ˆì•¡(TotalAmount) í•©ê³„ ì§‘ê³„
# ë°ì´í„° ê·¸ë£¹í™” ë° ì§‘ê³„
product_stats = df_sales.groupby('Product')[['Quantity', 'TotalAmount']].sum()  # í•©ê³„ ê³„ì‚°

# ë‹¨ê³„ 2: 'TotalAmount' ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ(descending) ì •ë ¬
# (ë¬¸ì œì—ì„œëŠ” Price ê¸°ì¤€ì´ë¼ê³  í–ˆìœ¼ë‚˜, ì§‘ê³„ëœ ë°ì´í„° í”„ë ˆì„ì—ëŠ” Price ì»¬ëŸ¼ì´ ì—†ê³  TotalAmountê°€ ë” ì ì ˆí•  ìˆ˜ ìˆìŒ. 
# í•˜ì§€ë§Œ ë¬¸ì œ ì˜ë„ê°€ 'ì§‘ê³„ í›„ ì´ íŒë§¤ê¸ˆì•¡ ìˆœ'ì´ë¼ë©´ TotalAmount ì‚¬ìš©, ë§Œì•½ 'ë‹¨ê°€'ë¼ë©´ ì›ë³¸ ì •ë³´ ë³‘í•© í•„ìš”.
# ì—¬ê¸°ì„œëŠ” 'ì´ íŒë§¤ ê¸ˆì•¡'ì„ ê¸°ì¤€ìœ¼ë¡œ ì •ë ¬í•˜ê² ìŠµë‹ˆë‹¤.)
# ë°ì´í„° ì •ë ¬
product_stats_sorted = product_stats.sort_values(by='TotalAmount', ascending=False)

print("\n== ìƒí’ˆë³„ íŒë§¤ëŸ‰ ë° ì´ íŒë§¤ ê¸ˆì•¡ (ë§¤ì¶œì•¡ ìˆœ ì •ë ¬) ==")
print(product_stats_sorted)
````
