## üåå ÏÇ∞Ï†êÎèÑ (Scatterplot)
Îëê Îç∞Ïù¥ÌÑ∞ Í∞ÑÏùò Í¥ÄÍ≥ÑÎ•º Ï†êÏúºÎ°ú ÌëúÌòÑÌï¥Ïöî.


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# Í∏∞Î≥∏ Îç∞Ïù¥ÌÑ∞ÏÖã Î°úÎìú
tips = sns.load_dataset('tips')

# Í∞ÑÎã®Ìïú ÏÇ∞Ï†êÎèÑ Í∑∏Î¶¨Í∏∞
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('ÌåÅÍ≥º Ï¥ù ÏßÄÎ∂àÏï°Ïùò Í¥ÄÍ≥Ñ')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  # numpy ÏßÅÏ†ë Í∞ÄÏ†∏Ïò§Í∏∞

# ÏòàÏ†ú Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
dates = pd.date_range('2021-01-01', periods=100)
prices = pd.Series(range(100)) + np.random.randn(100).cumsum()  # np.randomÏúºÎ°ú Î≥ÄÍ≤Ω
stock_data = pd.DataFrame({'Date': dates, 'Price': prices})

# ÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
sns.lineplot(x='Date', y='Price', data=stock_data)
plt.title('ÌïúÍ∏ÄÎ°ú Ï†úÎ™©ÏùÑ Ïì∞ÎäîÍ≤å Ïû¨Î∞åÏñ¥')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

## üìä ÎßâÎåÄ Í∑∏ÎûòÌîÑ (Barplot)
Î≤îÏ£º(Category)Î≥ÑÎ°ú Í∞íÏùò ÌÅ¨Í∏∞Î•º ÎßâÎåÄÎ°ú ÎπÑÍµêÌï¥Ïöî.


```python
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ÏÖã Î°úÎìú
tips = sns.load_dataset('tips')

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
sns.barplot(x='day', y='tip', data=tips, estimator=sum)
plt.title('ÎÇ†ÏßúÎ≥ÑÏ¥ùÌåÅÏù∏Í∞Ä ÏïÑÎßà Í∑∏Í±∞ÏùºÍ±∏')
plt.xlabel('ÎÇ†Ïßú')
plt.ylabel('Ï¥ùÌåÅ')
plt.show()
```

## üìà ÌûàÏä§ÌÜ†Í∑∏Îû® (Histplot)
Îç∞Ïù¥ÌÑ∞Í∞Ä Ïñ¥ÎñªÍ≤å Î∂ÑÌè¨Ìï¥ ÏûàÎäîÏßÄ(Ïñ¥ÎîîÏóê ÎßéÏù¥ Î™∞Î†§ÏûàÎäîÏßÄ) Î≥¥Ïó¨Ï§òÏöî.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ÏÖã Î°úÎìú
tips = sns.load_dataset('tips')

# ÌûàÏä§ÌÜ†Í∑∏Îû®Í≥º KDE Í∑∏Î¶¨Í∏∞
sns.histplot(tips['total_bill'], kde=True, bins=20)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

## üì¶ ÏÉÅÏûê Í∑∏Î¶º (Boxplot)
Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨ÏôÄ Ïù¥ÏÉÅÏπò(ÌäÄÎäî Í∞í)Î•º ÌïúÎààÏóê Î≥º Ïàò ÏûàÏñ¥Ïöî.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ÏÖã Î°úÎìú
tips = sns.load_dataset('tips')

# Î∞ïÏä§ÌîåÎ°Ø Í∑∏Î¶¨Í∏∞
sns.boxplot(x='sex', y='tip', data=tips)
plt.title('Tip Amount by Sex')
plt.xlabel('Sex')
plt.ylabel('Tip Amount')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Î∞îÏù¥Ïò¨Î¶∞ ÌîåÎ°Ø Í∑∏Î¶¨Í∏∞
sns.violinplot(x='sex', y='tip', data=tips)
plt.title('Tip Amount by Sex')
plt.xlabel('Sex')
plt.ylabel('Tip Amount')
plt.show()
```

## üß© ÌéòÏñ¥ÌîåÎ°Ø (Pairplot)
Ïó¨Îü¨ Îç∞Ïù¥ÌÑ∞ÎÅºÎ¶¨Ïùò Í¥ÄÍ≥ÑÎ•º Ìïú Î≤àÏóê Í∑∏ÎûòÌîÑÎ°ú Î≥¥Ïó¨Ï§òÏöî.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ÏÖã Î°úÎìú
iris = sns.load_dataset('iris')

# ÌéòÏñ¥ÌîåÎ°Ø Í∑∏Î¶¨Í∏∞
sns.pairplot(iris, hue='species')
plt.show()
```

## üåå ÏÇ∞Ï†êÎèÑ (Scatterplot)
Îëê Îç∞Ïù¥ÌÑ∞ Í∞ÑÏùò Í¥ÄÍ≥ÑÎ•º Ï†êÏúºÎ°ú ÌëúÌòÑÌï¥Ïöî.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ÏÖã Î°úÎìú
tips = sns.load_dataset('tips')

# ÏÇ∞Ï†êÎèÑ Í∑∏Î¶¨Í∏∞
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='time')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (ÌöåÏÇ¨Ïùò ÏõîÎ≥Ñ Ï†úÌíà ÌåêÎß§Îüâ)
data = {
    'January': [500, 600, 700, 400],
    'February': [550, 620, 680, 450],
    'March': [530, 610, 670, 420],
    'April': [580, 640, 710, 460],
    'May': [590, 660, 720, 470],
    'June': [620, 680, 750, 490]
}
companies = ['Company A', 'Company B', 'Company C', 'Company D']

df = pd.DataFrame(data, index=companies)

df
```

## üî• ÌûàÌä∏Îßµ (Heatmap)
Îç∞Ïù¥ÌÑ∞Ïùò ÌÅ¨Í∏∞Î•º ÏÉâÍπîÏùò ÏßÑÌïòÍ∏∞Î°ú ÌëúÌòÑÌï¥Ïöî. Ï£ºÎ°ú ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Î≥º Îïå Ïç®Ïöî.


```python
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Í∞Å ÌöåÏÇ¨Î≥Ñ ÏõîÎ≥Ñ Ï†úÌíà ÌåêÎß§Îüâ')
plt.show()
```

## üî• ÌûàÌä∏Îßµ (Heatmap)
Îç∞Ïù¥ÌÑ∞Ïùò ÌÅ¨Í∏∞Î•º ÏÉâÍπîÏùò ÏßÑÌïòÍ∏∞Î°ú ÌëúÌòÑÌï¥Ïöî. Ï£ºÎ°ú ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Î≥º Îïå Ïç®Ïöî.


```python
# Îç∞Ïù¥ÌÑ∞ Ï°∞Ïûë: Îã§ÏñëÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ (1, -1, 0Ïóê Í∞ÄÍπåÏö¥ Í∞íÏù¥ ÎÇòÏò§ÎèÑÎ°ù)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# ÏàòÏ†ïÎêú Îç∞Ïù¥ÌÑ∞ (Îã§ÏñëÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Ïú†ÎèÑ)
data = {
    'Stock Price': [120, 135, 150, 160, 145, 170, 185, 190],    # Ï£ºÍ∞Ä
    'Revenue': [50, 55, 65, 70, 67, 75, 80, 85],                # Îß§Ï∂ú
    'Profit': [30, 28, 25, 22, 21, 20, 18, 15],                # Ïù¥Ïùµ (Í∞êÏÜåÌïòÎäî Í∞í)
    'Expenses': [60, 58, 55, 53, 50, 48, 45, 43],              # ÎπÑÏö© (Í∞êÏÜå)
    'Debt': [80, 75, 70, 68, 72, 74, 78, 82],                  # Î∂ÄÏ±Ñ (Î≥ÄÎèô)
    'R&D Spending': [10, 10, 15, 15, 20, 20, 25, 25]           # Ïó∞Íµ¨Í∞úÎ∞úÎπÑ (ÏùºÏ†ïÌïú Ìå®ÌÑ¥)
}

# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
df = pd.DataFrame(data)

# ÏÉÅÍ¥ÄÍ≥ÑÏàò Í≥ÑÏÇ∞
corr_matrix = df.corr()

# Heatmap Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# ÌÉÄÏù¥ÌãÄ Ï∂îÍ∞Ä
plt.title('Correlation Between Financial Metrics')

# Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞
plt.show()

# ÏÉÅÍ¥ÄÍ≥ÑÏàò Ï∂úÎ†•
corr_matrix
```
