## [Ïã§Ïäµ 1] Îã§ÏñëÌïú Í∑∏ÎûòÌîÑ ÌôúÏö© 1

- Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ ÏÑ† Í∑∏ÎûòÌîÑ, ÎßâÎåÄ Í∑∏ÎûòÌîÑ, ÌûàÏä§ÌÜ†Í∑∏Îû®, ÏÇ∞Ï†êÎèÑÎ•º ÌïúÍ∫ºÎ≤àÏóê Í∑∏Î†§Î≥¥ÏÑ∏Ïöî.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
x = np.arange(1, 11)
y = np.random.randint(10, 100, 10)

data = np.random.normal(50, 15, 100)

plt.subplot(2, 2, 1)
plt.plot(x, y)                                    # ÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞

plt.subplot(2, 2, 2)
plt.bar(x, y)                                     # ÎßâÎåÄ Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞

plt.subplot(2, 2, 3)
plt.hist(data)

plt.subplot(2, 2, 4)
plt.scatter(x, y)                                 # ÏÇ∞Ï†êÎèÑ Í∑∏Î¶¨Í∏∞

plt.tight_layout()
plt.show()                                        # Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞
```

## [Ïã§Ïäµ 2] Îã§ÏñëÌïú Í∑∏ÎûòÌîÑ ÌôúÏö© 2

- Îã§Ïùå Îç∞Ïù¥ÌÑ∞Î•º Í∞ÄÏßÄÍ≥† Î¨∏Ï†úÎ•º Ìï¥Í≤∞ÌïòÏÑ∏Ïöî.
    
    <aside>
    üí°
    
    - **ÏõîÎ≥Ñ ÌèâÍ∑† Í∏∞Ïò® (¬∞C)**
        - `months`: ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
        - `avg_temperatures`: [-1, 0, 5, 10, 15, 20, 25, 23, 19, 12, 5, 0]
    - **ÏõîÎ≥Ñ Í∞ïÏàòÎüâ (mm)**
        - `rainfall`: [50, 40, 60, 80, 100, 120, 150, 140, 110, 90, 70, 60]
    - **ÌïôÏÉùÎì§Ïùò ÏãúÌóò Ï†êÏàò**
        - `math_scores`: [85, 90, 88, 76, 95, 89, 72, 84, 91, 87]
        - `english_scores`: [78, 82, 85, 80, 88, 90, 84, 79, 85, 83]
    </aside>

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']
avg_temperatures = [-1, 0, 5, 10, 15, 20, 25, 23, 19, 12, 5, 0]
rainfall = [50, 40, 60, 80, 100, 120, 150, 140, 110, 90, 70, 60]
math_scores = [85, 90, 88, 76, 95, 89, 72, 84, 91, 87]
english_scores = [78, 82, 85, 80, 88, 90, 84, 79, 85, 83]
```

1. ÏõîÎ≥Ñ ÌèâÍ∑† Í∏∞Ïò®ÏùÑ ÏãúÍ∞ÅÌôîÌïòÎäî ÏÑ† Í∑∏ÎûòÌîÑÎ•º Í∑∏Î†§Î≥¥ÏÑ∏Ïöî.
2. ÏõîÎ≥Ñ Í∞ïÏàòÎüâÏùÑ ÏãúÍ∞ÅÌôîÌïòÎäî ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î†§Î≥¥ÏÑ∏Ïöî.
3. ÏàòÌïôÍ≥º ÏòÅÏñ¥ Ï†êÏàòÏùò Î∂ÑÌè¨Î•º ÎπÑÍµêÌïòÎäî Î∞ïÏä§ÌîåÎ°Ø Í∑∏ÎûòÌîÑÎ•º Í∑∏Î†§Î≥¥ÏÑ∏Ïöî.
4. ÏàòÌïô Ï†êÏàò Î∂ÑÌè¨Î•º Î∂ÑÏÑùÌïòÎäî ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Í∑∏Î†§Î≥¥ÏÑ∏Ïöî.
5. ÏàòÌïô Ï†êÏàòÏôÄ ÏòÅÏñ¥ Ï†êÏàòÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Î∂ÑÏÑùÌïòÎäî ÏÇ∞Ï†êÎèÑÎ•º Í∑∏Î†§Î≥¥ÏÑ∏Ïöî.
6. 1~5Ïùò Î™®Îì† Í∑∏ÎûòÌîÑÎ•º ÌïòÎÇòÏùò FigureÏóê Î∞∞ÏπòÌï¥ Î≥¥ÏÑ∏Ïöî.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(12, 8))
# subplot (3, 2)
# Q1 ÏÑ†Í∑∏ÎûòÌîÑ
plt.subplot(3, 2, 1)
plt.plot(months, avg_temperatures, color="red", marker="o")  # ÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.xticks(rotation=30)
plt.title("ÏõîÎ≥Ñ ÌèâÍ∑† Í∏∞Ïò®")
plt.xlabel("Month")
plt.ylabel("Average Temperature")

# Q2 ÎßâÎåÄÍ∑∏ÎûòÌîÑ
plt.subplot(3, 2, 2)
plt.bar(months, rainfall, color="skyblue")        # ÎßâÎåÄ Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.xticks(months, [month + 1 for month in range(len(months))])
plt.title("ÏõîÎ≥Ñ Í∞ïÏàòÎüâ")
plt.xlabel("Month")
plt.ylabel("Rainfall")

# Q3 Î∞ïÏä§ÌîåÎ°Ø
plt.subplot(3, 2, 3)
plt.boxplot([math_scores, english_scores])
plt.title("ÏàòÌïôÍ≥º ÏòÅÏñ¥ Ï†êÏàò")
plt.xlabel("Subject")
plt.ylabel("Score")

# Q4 ÌûàÏä§ÌÜ†Í∑∏Îû®
plt.subplot(3, 2, 4)
plt.hist(math_scores, color="lightgreen", edgecolor="green")
plt.title("ÏàòÌïô Ï†êÏàò")
plt.xlabel("Score")
plt.ylabel("Frequency")

# Q5 ÏÇ∞Ìè¨ÎèÑ
plt.subplot(3, 2, 5)
plt.scatter(math_scores, english_scores, color="pink")  # ÏÇ∞Ï†êÎèÑ Í∑∏Î¶¨Í∏∞
plt.title("ÏàòÌïôÍ≥º ÏòÅÏñ¥ Ï†êÏàò")
plt.xlabel("Math Score")
plt.ylabel("English Score")

plt.tight_layout()
plt.show()                                        # Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞
```
