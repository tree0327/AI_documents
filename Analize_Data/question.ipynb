```python
import pandas as pd

# ì˜ˆì‹œ ìƒí’ˆ íŒë§¤ ë°ì´í„°
sales_data = {
    'OrderID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Monitor', 'Mouse', 'Keyboard', 'Laptop', 'Mouse', 'Monitor', 'Keyboard'],
    'Quantity': [1, 2, 1, 1, 2, 1, 1, 3, 1, 1],
    'Price': [1200, 25, 1200, 300, 25, 100, 1200, 25, 300, 100],
    'CustomerID': [1001, 1002, 1001, 1003, 1004, 1002, 1001, 1003, 1005, 1002],
    'OrderDate': ['2024-01-05', '2024-01-06', '2024-01-07', '2024-01-10', '2024-01-12', '2024-01-15', '2024-01-18', '2024-01-20', '2024-01-22', '2024-01-25']
}
df_sales = pd.DataFrame(sales_data)               # ë°ì´í„°í”„ë ˆì„(í‘œ) ìƒì„±
df_sales
```

# 1ë²ˆ

```python
# ì´ êµ¬ë§¤ê¸ˆì•¡ìœ¼ë¡œ ìƒˆë¡œìš´ ì»¬ëŸ¼ ë§Œë“¤ê¸°
df_sales['Total'] = df_sales['Quantity'] * df_sales['Price']

# ê³ ê°ë³„ ì´ êµ¬ë§¤ê¸ˆì•¡ ê³„ì‚°í•˜ê¸°
customer_total = df_sales.groupby('CustomerID')['Total'].sum()  # í•©ê³„ ê³„ì‚°

# ê²°ê³¼ ì¶œë ¥
customer_total
```

# 2ë²ˆ - Laptopì„ êµ¬ë§¤í•œ ê³ ê°ë“¤ì˜ í‰ê·  êµ¬ë§¤ ê¸ˆì•¡

```python
# 1. ë©íƒ‘ êµ¬ë§¤ì ë°ì´í„° êµ¬í•˜ê¸°
laptop_sales = df_sales[df_sales['Product'] == 'Laptop']
# laptop_sales

# 2. ë©íƒ‘ êµ¬ë§¤ìì˜ ì´ êµ¬ë§¤ê¸ˆì•¡ì˜ í‰ê· 
laptop_avg_purchase = laptop_sales['Total'].mean()  # í‰ê·  ê³„ì‚°
laptop_avg_purchase
```

# 3ë²ˆ - ê°€ì¥ ë§ì´íŒë§¤ëœ ìƒí’ˆì˜ ì´ë¦„ê³¼ íŒë§¤ëœ ì´ ìˆ˜ëŸ‰

```python
# ìƒí’ˆë³„ íŒë§¤ìˆ˜ëŸ‰ í•©ê³„ ê³„ì‚° (groupby ì‚¬ìš© reset_index ì‚¬ìš© ì•ˆí•¨)
product_sales = df_sales.groupby('Product')['Quantity'].sum()  # í•©ê³„ ê³„ì‚°
# product_sales
# idxmax(): ê°€ì¥ í° ê°’ì„ ê°€ì§„ ì¸ë±ìŠ¤ ë°˜í™˜ / max(): ê°€ì¥ í° ê°’ì„ ë°˜í™˜
best_product = product_sales.idxmax()
best_product_qt = product_sales.max()
print(f"ê°€ì¥ ë§ì´ íŒë§¤ëœ ìƒí’ˆ: {best_product} ({best_product_qt}ê°œ)")
```

# 4ë²ˆ - OrderDateê°€ 2024ë…„ 1ì›” 10ì¼ ì´í›„ì¸ ì£¼ë¬¸ë§Œ í•„í„°ë§í•˜ì—¬ ì¶œë ¥

```python
df_sales
# 1. ëŒ€ì†Œë¹„êµë¥¼ ìœ„í•´ OrderDate ì»¬ëŸ¼ì˜ ê°’ì„ datetime í˜•ì‹ìœ¼ë¡œ ë°˜í™˜ë°›ìŒ
df_sales['OrderDate'] = pd.to_datetime(df_sales['OrderDate'])
# df_sales.info()

# 2 OrderDate > '2024-01-10'ì¸ ë°ì´í„° ì¶œë ¥
filtered_sales = df_sales[df_sales['OrderDate'] > '2024-01-10']
filtered_sales
```

# 5ë²ˆ - Productë³„ íŒë§¤ëŸ‰ê³¼ ì´ íŒë§¤ ê¸ˆì•¡ì„ ê³„ì‚°í•œ í›„, ì´ë¥¼ ì´. ã…ã„´ë§¤ê¸ˆì•¡ ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬

```python
# ìƒí’ˆë³„ íŒë§¤ëŸ‰ê³¼ ì´ íŒë§¤ê¸ˆì•¡(aggí•¨ìˆ˜ ì‚¬ìš©)
product_stats = df_sales.groupby('Product').agg({'Quantity': 'sum', 'Total': 'sum'})
product_stats

# ì´ íŒë§¤ê¸ˆì•¡ ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
sorted_stats = product_stats.sort_values(by='Total', ascending=False)  # ê°’ ì •ë ¬í•˜ê¸°
sorted_stats
```
