### ğŸ‘Ÿ ì§•ê²€ë‹¤ë¦¬ ê±´ë„ˆê¸° (ê°„ê²© ì£¼ê¸°)

```python
address = 'ëŒ€í•œë¯¼êµ­ ì„œìš¸ì‹œ ì„œì´ˆêµ¬'
print(address[9:])      # ì„œì´ˆêµ¬ (9ë²ˆ ì¸ë±ìŠ¤ë¶€í„° ëê¹Œì§€)
print(address[-3:])     # ì„œì´ˆêµ¬ (ë’¤ì—ì„œ 3ë²ˆì§¸ ì¸ë±ìŠ¤ë¶€í„° ëê¹Œì§€)
print(address[1:12:4])  # í•œì„œì„œ (1ë²ˆ ì¸ë±ìŠ¤ë¶€í„° 12ë²ˆ ì¸ë±ìŠ¤ ì „ê¹Œì§€ ì¶”ì¶œí•˜ëŠ”ë° 4ê°œì”© ê±´ë„ˆë›°ë©´ì„œ ì¶”ì¶œ)
print(address[::-1])# êµ¬ì´ˆì„œ ì‹œìš¸ì„œ êµ­ë¯¼í•œëŒ€ (ìŠ¬ë¼ì´ì‹±ì„ ì´ìš©í•´ [::-1]ì„ ì£¼ë©´ ë¬¸ìì—´ ë’¤ì§‘ê¸°)
```
