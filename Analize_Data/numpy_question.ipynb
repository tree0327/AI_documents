````python
import numpy as np

# 5ê°œ ì§€ì ì˜ ì›”ë³„ ì‚¬ê³¼, ë°”ë‚˜ë‚˜, ì˜¤ë Œì§€ íŒë§¤ëŸ‰ (5x3x6 ë°°ì—´)
# 5ê°œì˜ ì§€ì , 3ê°œì˜ ê³¼ì¼(ì‚¬ê³¼, ë°”ë‚˜ë‚˜, ì˜¤ë Œì§€), 6ê°œì›”ì¹˜ ë°ì´í„°
sales_data = np.array([                           # ë°°ì—´ ìƒì„±
    # ì§€ì  1
    [[120, 150, 200, 230, 180, 210],  # ì‚¬ê³¼ íŒë§¤ëŸ‰
     [100, 140, 170, 190, 160, 180],  # ë°”ë‚˜ë‚˜ íŒë§¤ëŸ‰
     [90, 120, 160, 140, 130, 150]],  # ì˜¤ë Œì§€ íŒë§¤ëŸ‰

    # ì§€ì  2
    [[80, 90, 110, 140, 130, 120],
     [60, 70, 90, 100, 110, 95],
     [50, 80, 100, 110, 90, 85]],

    # ì§€ì  3
    [[130, 140, 160, 180, 170, 190],
     [110, 120, 130, 150, 140, 160],
     [100, 110, 120, 130, 120, 140]],

    # ì§€ì  4
    [[100, 130, 150, 170, 160, 180],
     [90, 100, 110, 120, 130, 125],
     [80, 110, 140, 130, 120, 110]],

    # ì§€ì  5
    [[90, 110, 130, 140, 120, 130],
     [70, 85, 100, 110, 105, 95],
     [60, 90, 110, 100, 90, 95]]
])
````

````python
# 1. ê° ì§€ì ì—ì„œ 6ê°œì›” ë™ì•ˆ íŒë§¤ëœ ì‚¬ê³¼, ë°”ë‚˜ë‚˜, ì˜¤ë Œì§€ì˜ ì›”ë³„ í‰ê·  íŒë§¤ëŸ‰ì„ ê³„ì‚°í•˜ì„¸ìš”.
print(sales_data.shape)  # (5, 3, 6)  => axis=0 5ê°œ ì§€ì , axis=1 3ê°œ ê³¼ì¼, axis=2 6ê°œì›”ì¹˜ ì›”ë³„ ë°ì´í„°

avg_sales = np.mean(sales_data, axis=2)
print(avg_sales)
````

````python
# 2. ê° ì›”ë§ˆë‹¤ ì „ì²´ ì§€ì ì—ì„œ ê°€ì¥ ë§ì´ íŒë§¤ëœ ê³¼ì¼ì„ ì°¾ì•„ íŒë§¤ëŸ‰ê³¼ ê³¼ì¼ëª…ì„ ì¶œë ¥í•˜ì„¸ìš”.
total_sales_per_month = np.sum(sales_data, axis=0)
print(total_sales_per_month)
print(total_sales_per_month.shape)                # ë°ì´í„°ì˜ í¬ê¸°(í–‰, ì—´) í™•ì¸

max_sales_per_month = np.max(total_sales_per_month, axis=0)    # ê° ì›”ë³„ ìµœëŒ€ íŒë§¤ëŸ‰
max_sales_indicies = np.argmax(total_sales_per_month, axis=0)  # ìµœëŒ€ íŒë§¤ëŸ‰ì´ ê¸°ë¡ëœ ê³¼ì¼ ì¸ë±ìŠ¤
print(max_sales_per_month)
print(max_sales_indicies)

fruits = np.array(['ì‚¬ê³¼', 'ë°”ë‚˜ë‚˜', 'ì˜¤ë Œì§€'])           # ë°°ì—´ ìƒì„±
print(fruits[max_sales_indicies])
````

3. ì „ì²´ ì§€ì ì—ì„œ ê³¼ì¼ë³„ ì´ íŒë§¤ëŸ‰ì„ êµ¬í•˜ê³ , ê°€ì¥ ë§ì´ íŒë§¤ëœ ê³¼ì¼ì„ ì°¾ëŠ”ë‹¤.

````python
total_sales_per_fruit = np.sum(sales_data, axis = (0, 2))
print(total_sales_per_fruit)

fruits = np.array(['ì‚¬ê³¼', 'ë°”ë‚˜ë‚˜', 'ì˜¤ë Œì§€'])           # ë°°ì—´ ìƒì„±
most_fruit = np.argmax(total_sales_per_fruit)
print(most_fruit)
print(fruits[most_fruit])
````

````python
# 4. ê° ì§€ì ì—ì„œ íŒë§¤ëŸ‰ì´ 150ê°œ ì´ìƒì¸ ë°ì´í„°ë¥¼ ì¶”ì¶œí•˜ê³  ê·¸ì— í•´ë‹¹í•˜ëŠ” ì§€ì ê³¼ ê³¼ì¼ì„ ì¶œë ¥í•˜ì„¸ìš”.
fruits = np.array(['ì‚¬ê³¼', 'ë°”ë‚˜ë‚˜', 'ì˜¤ë Œì§€'])           # ë°°ì—´ ìƒì„±

mask = sales_data >= 150
locations = np.argwhere(mask)                     # ì´ë¦„ìœ¼ë¡œ ë°ì´í„° ì„ íƒ
print(locations)                                  # ì´ë¦„ìœ¼ë¡œ ë°ì´í„° ì„ íƒ

for loc in locations:                             # ì´ë¦„ìœ¼ë¡œ ë°ì´í„° ì„ íƒ
    site, fruit, month = loc                      # ì´ë¦„ìœ¼ë¡œ ë°ì´í„° ì„ íƒ
    print(f"ì§€ì  : {site+1}, ê³¼ì¼ëª… : {fruits[fruit]}, {month+1}ì›” : {sales_data[site, fruit, month]}")
````
