```python
import random
```

```python
# 0 이상이면서 1 미만인 부동소수점의 난수를 생성
num = random.random()
print(num)
```

```python
# 1이상 10이하의 정수 난수 생성
num2 = random.randint(1, 10)
print(num2)
```

```python
# 주어진 시퀀스 자료형에서 임의의 요소를 하나 반환합니다.
fruits = ["apple", "banana", "orange", "grape"]
random_fruits = random.choice(fruits)

print(random_fruits)
```

```python
# 리스트의 요소를 임의의 순서로 섞는다.
cards = list(map(str, range(2, 11))) + list("AJQK")
random.shuffle(cards)
print(cards)
```

```python
# 리스트에서 중복되지 않는 3개의 요소를 선택해온다.
numbers = list(range(1, 11))
random_selection = random.sample(numbers, 3)
print(random_selection)
```
