## 1. 클래스(Class)와 객체(Object): 붕어빵 틀

프로그래밍에서 가장 중요한 개념 중 하나입니다.

**비유**:
- **클래스 (Class)**: **붕어빵 틀 (설계도)**. 팥을 넣을지 크림을 넣을지 기능이 정의되어 있음.
- **객체 (Object)**: **실제 붕어빵**. 틀에서 찍어내어 실제로 먹을 수 있는 것.

클래스를 하나 잘 만들어두면, 객체(붕어빵)를 수백 개든 쉽게 만들 수 있습니다.

```python
# 1. 붕어빵 틀 만들기 (클래스 정의)
class Bungeoppang:
    def __init__(self, content):
        self.content = content  # 속성 (팥, 슈크림 등)

    def eat(self):
        print(f"{self.content} 붕어빵을 냠냠 먹습니다. 맛있다!")

# 2. 붕어빵 찍어내기 (객체 생성)
fish1 = Bungeoppang("팥")
fish2 = Bungeoppang("슈크림")

# 3. 먹어보기 (메서드 호출)
fish1.eat()
fish2.eat()
```

## 2. 예외 처리 (Try-Except): 안전 그물망

프로그램을 돌리다 보면 에러가 나서 툭 꺼질 때가 있습니다. 이걸 방지하는 것이 **예외 처리**입니다.

**비유**: **서커스 안전 그물**
- 공중곡예를 하다가 떨어져도(에러 발생), 그물(Except)이 있어서 죽지 않고 다시 일어설 수 있습니다.
- 에러가 나도 프로그램이 종료되지 않고, "문제가 생겼지만 계속 할게요~"라고 처리하는 기술입니다.

```python
def divide_number(a, b):
    try:
        result = a / b
        print(f"결과: {result}")
    except ZeroDivisionError:
        print("⚠️ 에러! 0으로 나눌 수는 없습니다. (안전 그물 작동)")
    except Exception as e:
        print(f"⚠️ 알 수 없는 에러가 발생했습니다: {e}")

# 정상 실행
divide_number(10, 2)

# 에러 유발 (0으로 나누기)
divide_number(10, 0)

print("프로그램이 죽지 않고 잘 마무리되었습니다.")
```
