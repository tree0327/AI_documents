```python
#str: 시퀀스 자료형 중 하나
#시퀀스 자료형: 연속해서 값을 갖는 자료형
#str literal: 작은(큰)따옴표를 작성해서 텍스트의 범위를 나타낸다.
x = '이것은 문자열 리터럴입니다.' # <- 문자열 리터럴
y = "이것은 문자열 리터럴입니다." # <- 문자열 리터럴
print(x)
```

```python
#만약에 작은 따옴표를 텍스트 자체로 사용하고 싶다면
# 1. 역슬래시 \를 작은 따옴표 앞에 붙여서 쓰면 된다.
# 2. 큰따옴표(") 안에 사용할떄는 '를 그대로 사용해도 된다.
text1 = 'don\'t do that'
text2 = "don't do that"
print(text1)
```

```python
# 문자열 리터럴 안에 \n은 줄바꿈을 의미
text3 = '첫번째 문장\n두번째 문장'
print(text3)
```

```python
# 경우에 따라 \n을 문자열에 담고 싶을 때
# 1. raw str
# 2. \\ 를 이용해 \문자 그대로 출력
text4 = r'c:\name'
text5 = 'c:\\name'
print(text4)
print(text5)
```

```python
#삼중 따옴표를 사용하여 문자열 리터럴 생성
text6 = """동해물과 백두산이
마르고 닳도록
끝"""
print(text6)
```

```python
# 문자열에 대해서도 *, +를 할 수 있다.
# *: 앞에 오는 문자열에 대해서 뒤에 오는 숫자만큼 반복해서 이어준다.
# +: 더하기 연산자 앞/뒤로 있는 문자열을 합쳐서 이어준다.
text7 = '너무 '
text8 = '파이썬 공부가 즐겁습니다.'
print(text7 * 10)
print(text7 + text8)
print(text7 * 10 + text8)
```

```python
# index: 0부터 순서대로 1씩 숫자가 증가 -1은 제일 뒤에서 부터 시작

text9 = 'Python'
print(text9[0])
```

```python
# slicing: 범위로써 인덱스 위치에 있는 요소를 꺼내온다.
print(text9[1:4])
```

```python
print(text9[0:2])
```

```python
print(text9[:2])
```

```python
print(text9[:])
```

```python
print(text9[::2])
```

```python
# 슬라이싱을 통해 요소를 변경하려는 경우
# 좌항의 요소 개수와 우항의 요소 개수가 같아야 한다.
# 문자열(str)의 경우에는 immutable이기 떄문에 변경 불가
```

```python
text10 = "Aa" + text9[2:]
print(text10)

# "Aa" + "thon"
# "Aathon"
```

```python
# len() : 소괄호 안에 입력한 시퀀스의 크기를 구해준다.
len(text10)
```
