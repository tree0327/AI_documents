```python
# 사칙연산을 하는 간단한 함수

# 함수에 대한 정의부분
def plus(a, b):
    """ 두 개의 숫자를 더해서 그 값을 반환합니다."""
    return a + b;

def minus(a, b):
    """ 두 개의 숫자를 뺴서 그 값을 반환합니다."""
    return a - b;

def multi(a, b):
    """ 두 개의 숫자를 곱해서 그 값을 반환합니다."""
    return a * b;

def divide(a, b):
    """ 두 개의 숫자를 나눠서 그 값을 반환합니다."""
    return a / b;
```

```python
# 함수의 호출
# 함수의 이름: plus, minus, multi, divide
plus_result = plus(5, 3)
# plus_result = 8 -> 반환값으로 return 오른편에 있는 값이 호출하는 부분에 대치가 된다.
print(plus_result)
```

```python
minus_result = minus(5, 3)
print(minus_result)
```

```python
multi_result = multi(a=5, b=3) # keyword argument로 전달
print(multi_result)
```

```python
divide_result = divide(5, b=3) # positional과 keyword 방식을 섞어 쓸 경우 positional을 왼편에 사용
print(divide_result)
```
