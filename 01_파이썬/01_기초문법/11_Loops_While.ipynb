```python
# while문
# 반복문
# for문이 요소가 있을 때 하나씩 꺼내서 하나 꺼낼 때마다 코드를 반복했다면
# while문은 우측에 위치하는 조건문이 True일 때 코드블록을 실행하고 다시 조건문으로 돌아와 조건문이 True라면 코드블럭을 실행,
# 조건문의 최종값이 False라면 while문을 빠져나간다.

i = 0 # 초기값
while i < 10: # 조건식
    print("Hello world")
    i+=1 # 증감식
```

```python
scores = [98, 89, 28]
```

```python
i = 0
while i < len(scores):
    print(scores[i])
    i += 1
```

```python
# break는 반복문에서 break를 만나게 되면 제일 가까이 있는 반복문을 빠져나간다.

scores_2 = [98, 89, 28, 100]

idx_2 = 0
while idx_2 < len(scores_2):
    print(scores_2[idx_2])
    if scores_2[idx_2] < 90:
        break
    idx_2 += 1
```

```python
# continue: 반복문에서 만나게 됐을 때, while문에서는 조건식으로 이동합니다.
# for문에서는 iterable(한 번에 하나씩 요소를 꺼낼 수 있는 객체) 안에 남아있는 욧호가 있는지 체크하는 곳으로 이동

i = 0
while i < 100: # 조건식
    i += 1 # 증감식
    if i % 2 == 0: # 2의 배수일 때 (i를 2로 나누었을 때 나머지가 0인 경우) if문
        continue
    print(i)
```
